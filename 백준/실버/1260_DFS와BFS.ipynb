- 그래프를 DFS로 탐색한 결과와 BFS로 탐색한 결과를 각각 출력하는 프로그램
- 정점번호 1번부터 V번까지

```python
# 120ms
# dfs의 경우 방문 배열 생성 위치 주의하기
# 방문여부 탐색 위치 주의하기
# 낮은 숫자부터 방문할 때 정렬 방식 잘 기억하기
# 인접 정점 배열 받는법 잘 기억하기

def dfs(n) :
    dfs_ans.append(n)
    v[n] = 1

    for w in adj[n] :
        if v[w] == 0 :
            dfs(w)


    
def bfs(n) :
    q = []
    v = [0] * (V+1)

    q.append(n)
    bfs_ans.append(n)
    v[n] = 1

    while q :
        c = q.pop(0)
        for w in adj[c] :
            if v[w] == 0 :
                q.append(w)
                bfs_ans.append(w)
                v[w] = 1
    


# 정점개수, 간선개수, 시작점
V,E,S = map(int,input().split())
adj = [[] for _ in range(V+1)]      # 이거 만드는거 주의ㅡ
for _ in range(E) :
    s,e = map(int,input().split())
    adj[s].append(e)
    adj[e].append(s)

# 오름차순 정렬(작은 것부터 방문하기 위해)
for i in range(1,V+1) :
    adj[i].sort()

# 필요한거 만들기
v = [0] * (V+1)
dfs_ans = []
dfs(S)


bfs_ans = []
bfs(S)


print(*dfs_ans)
print(*bfs_ans)
```
