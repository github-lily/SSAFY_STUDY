### 문제설명
- 배추들이 모여있는 곳에는 배추 흰 지렁이가 한마리만 있으면 됨
- 필요한 배추 흰지렁이 마리수 구하기

### 입력값
- 첫줄 : 테스트 케이스의 개수
- 그 다음 : 배추밭의 가로길이 M, 세로 길이 N, 배추가 심어져있는 위치의 개수 K
- 배추의 위치 X,Y

### 출력값
- 필요한 배추흰지렁이 마리수

````python
# 꼼수 도전해봄
# 이전것이랑만 비교하기때문에 정확하지 않음


T = int(input())
for _ in range(T) :
    M, N, K = map(int,input().split())
    cnt = 1
    lst = []

    for _ in range(K) :
        X,Y = map(int,input().split())
        lst.append((X,Y))
    
    lst.sort()
    
    for i in range(1,K) :
        x1 = lst[i-1][0]
        x2 = lst[i][0]
        y1 = lst[i-1][1]
        y2 = lst[i][1]
        
        if (x1 == x2 and abs(y1-y2)<2) or (abs(x1-x2)<2 and y1==y2) :
            continue
        else :
            cnt += 1
    
    print(cnt)
````

## 재귀함수호출 한도를 설정하는 기능

```python
import sys

```

- 기본적인 파이썬의 재귀 호출 한도는 1000회(스택 오버플로우를 방지하기 위해 존재)
- 하지만 복잡한 문제에서는 수천번의 재귀가 필요할 수 있음
- 이 문제에서도 배추밭의 최대 크기는 50x50으로 수천번의 재귀가 호출될 수 있어 위 함수가 필요함
- 보통 10000정도면 적당. 너무 큰 값은 메모리가 부족해질 수 있음



````python
import sys
sys.setrecursionlimit(10000)
# 재귀함수호출 한도를 설정하는 기능
# 이거 없으면 RecursionError 에러 발생

dr = [0,1,0,-1]
dc = [1,0,-1,0]

def dfs(x,y) :
    v[x][y] = 1

    for i in range(4) :
        nr, nc = x+dr[i], y+dc[i]
        if 0<=nr<N and 0<=nc<M and arr[nr][nc] == 1 and v[nr][nc] == 0 :
            dfs(nr,nc)


T = int(input())
for _ in range(T) :
    M, N, K = map(int,input().split())

    # 필요 변수 선언
    ans = 0
    arr = [[0] * M for _ in range(N)]
    v = [[0] * M for _ in range(N)]

    # 배추 심기
    for _ in range(K) :
        X,Y = map(int,input().split())
        arr[Y][X] = 1


    # 배추밭 탐색
    for r in range(N) :
        for c in range(M) :
            if arr[r][c] == 1 and v[r][c] == 0 :
                dfs(r,c)
                ans += 1

    print(ans)
````
