- 아군은 W, 적군은 B
- 같은 팀의 병사 N명이 상하좌우로 뭉쳐있으면 N^2 의 위력을 냄
- 아군과 적군 병사의 위력의 합을 출력

```python
# dfs 실패
# 다른 영역을 만날때마다 값을 제곱하고 더해서 정확히 안나옴

def dfs(sr,sc,team,cnt) :
    global W_power, B_power

    if arr[sr][sc] != team :
        if team == 'W' :
            W_power += cnt*cnt
        else :
            B_power += cnt*cnt
        return
    
    for dr,dc in ((-1,0), (0,1),(1,0),(0,-1)) :
        nr,nc = sr+dr, sc+dc
        if 0<= nr<M and 0<= nc < N and arr[nr][nc] == team and v[nr][nc] == 0 :
            arr[nr][nc] = 0
            v[nr][nc] = 1
            dfs(nr,nc,team,cnt+1)
    

N,M = map(int,input().split())
arr = [list(input()) for _ in range(M)]
v = [[0]*N for _ in range(M)]
W_power = 0
B_power = 0

for r in range(M) :
    for c in range(N) :
        if arr[r][c] != 0 :
            dfs(r,c,arr[r][c],1)

print(W_power, B_power)
```

```python
# bfs 성공 (116ms)
# 방문한 곳은 0으로 바꿔서 bfs가 다시 탐색하지 않도록 함


def bfs(sr,sc,team) :
    global cnt
    q = []

    q.append((sr,sc))
    v[sr][sc] = 1
    arr[sr][sc] = 0

    while q :
        cr,cc = q.pop(0)
        for dr,dc in ((-1,0), (0,1),(1,0),(0,-1)) :
            nr,nc = cr+dr, cc+dc
            if 0<= nr<M and 0<= nc < N and arr[nr][nc] == team and v[nr][nc] == 0 :
                v[nr][nc] = 1
                cnt += 1
                arr[nr][nc] = 0     #배열에서도 방문한 곳 0 표시
                q.append((nr,nc))
    return cnt, team                #arr을 0으로 바꾸기 때문에 team을 함께 넘겨줘야 함



N,M = map(int,input().split())
arr = [list(input()) for _ in range(M)]
v = [[0]*N for _ in range(M)]
W_power = 0
B_power = 0

for r in range(M) :
    for c in range(N) :
        if arr[r][c] != 0 :
            cnt = 1
            sum_val, team = bfs(r,c,arr[r][c])
            if team == 'W' :
                W_power += (sum_val*sum_val)
            else :
                B_power += (sum_val*sum_val)


print(W_power, B_power)
```

```python
# 참고 코드
# https://covenant.tistory.com/147
# 방문배열을 이용해 방문 안한 곳을 탐색하는걸로도 이용할 수 있음(더 편리)


from collections import deque
 
def BFS():
    q.append((i, j))
    check[i][j] = 1
    t = 1
    while q:
        x, y = q.popleft()
        for k in range(4):
            nx, ny = x + dx[k], y + dy[k]
            if 0 <= nx < M and 0 <= ny < N and MAP[nx][ny] == MAP[x][y] and check[nx][ny] == 0:
                check[nx][ny] = 1
                q.append((nx, ny))
                t += 1
    return t
 
# N: 가로 M : 세로
N, M = map(int, input().split())
MAP = [list(input()) for _ in range(M)]
check = [[0] * N for _ in range(M)]
 
dx,dy=[1,-1,0,0],[0,0,1,-1]
q = deque()
B, W = 0, 0
 
for i in range(M):
    for j in range(N):
        if check[i][j] == 0:
            ans = BFS()
            if MAP[i][j] == 'W': W += ans ** 2
            else: B += ans ** 2
 
print(W, B)
```
