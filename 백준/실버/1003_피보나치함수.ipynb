```python
# 시간초과!

def fibo(n) :
    global zero, one

    if n == 0 :
        zero += 1
        return 0

    if n == 1 :
        one += 1
        return 1
    
    return fibo(n-2) + fibo(n-1)


T = int(input())
for _ in range(T) :
    N = int(input())
    zero = one = 0

    print(zero, one)

```

```python
# 110736KB, 108ms(PyPy3)

def fibo(n) :

    if n == 0 :
        return 0

    if n == 1 or n == 2 :
        return 1
    
    if dp[n] != 0 :
        return dp[n]
    
    dp[n] = fibo(n-2) + fibo(n-1)
    return dp[n]


T = int(input())
for _ in range(T) :
    N = int(input())
    dp = [0]*41
    dp[1] = dp[2] = 1

    if N == 0 :
        print(1, 0)

    else :
        fibo(N)

        zero = dp[N-1]
        one = dp[N]


        print(zero, one)
```

```python
# 코드 개선

# 109544KB, 96ms(PyPy3)


T = int(input())
for _ in range(T) :
    N = int(input())
    dp = [0]*41
    dp[1] = dp[2] = 1

    if N == 0 :
        print(1, 0)

    else :
        for i in range(3,41) :
            dp[i] = dp[i-1] + dp[i-2]

        zero = dp[N-1]
        one = dp[N]


        print(zero, one)
```

```python
# 코드 개선

# 109544KB, 104ms(PyPy3)
# for문을 필요한 정도만 돌도록 수정했으나 유의미한 차이 없음
# 이유
# 1. N의 범위가 작아서 반복 횟수 차이가 크지 않기 때문
# 2. 피보나치 수열은 단순 덧셈 연산으로 이루어져 연산 자체가 가벼움
# 3. PyPy3는 JIT 컴파일러를 이용해 반복문을 효율적으로 처리함


T = int(input())
for _ in range(T) :
    N = int(input())
    dp = [0]*41
    dp[1] = dp[2] = 1

    if N == 0 :
        print(1, 0)

    else :
        for i in range(3,N+1) :
            dp[i] = dp[i-1] + dp[i-2]

        zero = dp[N-1]
        one = dp[N]


        print(zero, one)
```
