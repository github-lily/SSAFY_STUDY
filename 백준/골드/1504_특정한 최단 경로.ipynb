### 문제 설명
- 무향 그래프
- 1번에서 N번 정점으로 이동
- 임의로 주어진 두 정점은 반드시 통과해야 한다
- 빙문했던 정점과 간선을 다시 방문할 수 있음(단, 최단경로일 때)

### 입력
1. 정점의 개수 N, 간선의 개수 E (2 ≤ N ≤ 800, 0 ≤ E ≤ 200,000) 
2. a,b,c (a,b는 연결 관계, 길이 c)
3. u,v 반드시 거쳐야 하는 정점 2개

### 출력
- 두개의 정점을 지나는 최단 경로의 길이
- 도달할 수 없을 땐 -1

```python
# 2차 시도
# path 경로를 들고다니며 최종에만 확인인
# 답은 나오지만 메모리 초과

import heapq

INF = int(1e9)
N,E = map(int,input().split())
connect = [[] for _ in range(N+1)]      # 정점번호 1부터 시작
ans = INF

for _ in range(E) :
    a,b,c = map(int,input().split())
    connect[a].append((b,c))
    connect[b].append((a,c))
    
# 방문해야하는 포인트
p1,p2 = map(int,input().split())



def dikjstra(start, p1, p2) :
    global ans

    q = []
    path = []
    heapq.heappush(q,(0,start,path+[start]))   # 현재까지의 거리, 노드, 경로

    while q :
        now_distance, node, path = heapq.heappop(q)

        for next, next_distance in connect[node] :
            distance = next_distance + now_distance
            # 노드가 목적지에 도착했고, 경로에 p1,p2가 모두 있다면
            if next == N :
                if p1 in path and p2 in path :
                    if ans > distance :
                        ans = distance
                        return
                        print(f'N_path : {path}')
                        print(f'N_next : {next}')
                        print(f'N_distance : {distance}')

            # 목적지가 아니라면 최단경로 저장
            else :
                heapq.heappush(q,(distance,next,path+[next]))



dikjstra(1,p1,p2)

print(ans)
```

```python
# 플로이드 워셜 방법
# 플로이드워셜의 경우 시간복잡도가 O(N^3)
# 이 문제의 경우 (2 ≤ N ≤ 800, 0 ≤ E ≤ 200,000) 이므로
# N = 800일 경우 512*10^6번의 연산이 필요하므로 사용 불가함

INF = int(1e9)
N, E = map(int, input().split())
dist = [[INF] * (N + 1) for _ in range(N + 1)]

# 자기 자신으로 가는 비용은 0으로 초기화
for i in range(1, N + 1):
    dist[i][i] = 0

# 간선 정보 입력
for _ in range(E):
    a, b, c = map(int, input().split())
    dist[a][b] = c
    dist[b][a] = c

# 플로이드-워셜 알고리즘
for k in range(1, N + 1):
    for i in range(1, N + 1):
        for j in range(1, N + 1):
            dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])

# 방문해야 하는 포인트
p1, p2 = map(int, input().split())

# 1 -> p1 -> p2 -> N 경로와 1 -> p2 -> p1 -> N 경로 중 최단 경로 계산
path1 = dist[1][p1] + dist[p1][p2] + dist[p2][N]
path2 = dist[1][p2] + dist[p2][p1] + dist[p1][N]

result = min(path1, path2)

# 도달할 수 없는 경우
if result >= INF:
    print(-1)
else:
    print(result)
```
