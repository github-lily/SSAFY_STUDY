```python
def can_attach(paper, x, y, size):
    # 색종이를 해당 크기만큼 붙일 수 있는지 확인하는 함수
    if x + size > 10 or y + size > 10:  # 격자를 벗어나면 False
        return False
    for i in range(size):
        for j in range(size):
            if paper[x + i][y + j] == 0:  # 색종이를 붙일 곳에 1이 없으면 False
                return False
    return True

def attach(paper, x, y, size, value):
    # 색종이를 붙이거나(0으로 변경) 떼는(1로 변경) 함수
    for i in range(size):
        for j in range(size):
            paper[x + i][y + j] = value

def solve(paper, x, y, used, remains):
    global result
    if x == 10:  # 모든 행을 다 처리한 경우
        result = min(result, used)
        return
    if y == 10:  # 현재 행의 끝까지 다 본 경우, 다음 행으로 이동
        solve(paper, x + 1, 0, used, remains)
        return
    if paper[x][y] == 0:  # 이미 0으로 덮인 부분은 넘어감
        solve(paper, x, y + 1, used, remains)
        return

    # 색종이를 크기별로 시도 (5x5부터 1x1까지)
    for size in range(5, 0, -1):
        if remains[size] > 0 and can_attach(paper, x, y, size):  # 색종이를 붙일 수 있다면
            attach(paper, x, y, size, 0)  # 색종이 붙임
            remains[size] -= 1  # 남은 색종이 수 감소
            solve(paper, x, y + size, used + 1, remains)  # 다음으로 넘어감
            attach(paper, x, y, size, 1)  # 다시 색종이 뗌 (백트래킹)
            remains[size] += 1  # 색종이 수 복구

# 입력 받기
paper = [list(map(int, input().split())) for _ in range(10)]
remains = [0, 5, 5, 5, 5, 5]  # 색종이 크기별로 5장씩
result = float('inf')  # 결과를 무한대로 초기화

solve(paper, 0, 0, 0, remains)
print(result if result != float('inf') else -1)  # 결과 출력
```

## idea 정리
### 1. 값 입력 받고, 변수 지정(정답변수)
### 2. 백트래킹을 통해 최소 색종이 개수 탐색
### 3. 색종이를 붙일 수 있는지 확인하는 함수 생성
### 4. 색종이를 붙인 곳은 헷갈리지 않게 0으로 값 변경

```python

# 색종이를 붙일 수 있는지 유무를 판단하는 함수
def possible(i,j,size) :
    if i+size > 10 or j+size > 10 :       # 범위를 벗어나면 리턴
        return False
    for r in range(size) :
        for c in range(size) :
            if paper[i+r][j+c] == 0 :       # 색종이로 덮을 곳이 0이라면 리턴
                return False                
    return True


# 색종이를 붙인 곳은 0으로, 뗀 곳은 1로 재설정해주는 함수
def change(i,j,size,value) :    
    for r in range(size) :
        for c in range(size) :
            paper[i+r][j+c] = value


def dfs(i,j,remains,used) :
    global ans
    if i == 10 :        #모든 행을 다 본 경우 최소 개수 구하기
        ans = min(ans,used)
        return
    
    if j == 10 :        #모든 열을 다 본 경우, 다음 행으로 이동
        dfs(i+1,0,remains,used)
        return
    

    if paper[i][j] == 0 :   # 이미 덮인 곳은 지나감
        dfs(i,j+1,remains,used)
        return
    
    for size in range(5,0,-1) :
        if remains[size] > 0 and possible(i,j,size) :   # 남아있는 종이가 있고, 사이즈가 맞는다면
            change(i,j,size,0)                          # 색종이 붙여주고(0으로 바꿈)
            remains[size] -=1                           # 사용한 색종이 줄여주고
            dfs(i,j+size,remains,used+1)                # 다음 함수 호출(붙인만큼 건너뛰고 시작, 사용횟수 하나 늘려줌)
            remains[size] += 1                          # 원상복구
            change(i,j,size,1)                          # 색종이도 떼줌

            
    


paper = [list(map(int,input().split())) for _ in range(10)]
remains = [0,5,5,5,5,5]         # 색종이는 5장씩
ans = float('inf')


dfs(0,0,remains,0)
if ans == float('inf') :
    print(-1)
else :
    print(ans)
```
