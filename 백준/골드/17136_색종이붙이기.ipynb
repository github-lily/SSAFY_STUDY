```python
def is_valid(paper, x, y, size):
    if x + size > 10 or y + size > 10:
        return False
    for i in range(size):
        for j in range(size):
            if paper[x + i][y + j] == 0:
                return False
    return True

def attach(paper, x, y, size, state):
    for i in range(size):
        for j in range(size):
            paper[x + i][y + j] = state

def dfs(paper, x, y, count, remains):
    global result
    if count >= result:
        return
    if x == 10:
        result = min(result, count)
        return
    if y == 10:
        dfs(paper, x + 1, 0, count, remains)
        return
    if paper[x][y] == 0:
        dfs(paper, x, y + 1, count, remains)
        return

    for size in range(5, 0, -1):
        if remains[size] > 0 and is_valid(paper, x, y, size):
            attach(paper, x, y, size, 0)
            remains[size] -= 1
            dfs(paper, x, y + size, count + 1, remains)
            attach(paper, x, y, size, 1)
            remains[size] += 1

paper = [list(map(int, input().split())) for _ in range(10)]
remains = [0, 5, 5, 5, 5, 5]
result = float('inf')

dfs(paper, 0, 0, 0, remains)
print(result if result != float('inf') else -1)
```
