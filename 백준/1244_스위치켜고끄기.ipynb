```python
# 1차시도
# tc는 맞지만 답을 틀림
# 한 줄에 20개씩 출력한다는 조건 빼먹음
# 인덱스 1부터 시작하니까 범위 확인할 때에도 1부터로 생각해야함
# 배수도 N+1까지 봐야함


'''
- 1부터 시작
- 남학생(1)은 받은 번호의 배수 스위치 상태를 바꾼다
- 여학생(2)은 받은 번호의 상태를 바꾸고, 좌우 대칭의 값이 같으면 다른 걸 만날 때까지 상태를 바꾼다

'''


N = int(input())            # 스위치 개수
arr = list(map(int,input().split()))
arr = [0] + arr     #인덱스 맞춰주기 위해 값 추가
M = int(input())            # 학생 수
for _ in range(M) :
    sex, num = map(int,input().split())

    # 남학생일 때
    if sex == 1 :
        for i in range(1,N) :
            if num * i > N :
                break
            else :
                arr[num*i] = 1-arr[num*i]
    # 여학생일 때
    else :
        arr[num] = 1 - arr[num]
        for i in range(1,N//2+1) :
            left = num-i
            right = num+i
            if 0<= left < N and 0<= right < N :
                if arr[left] == arr[right] :
                    arr[left] = arr[right] = 1-arr[left]
                else :
                    break       #대칭이 아니면 중단
            else :      #범위 밖이면 중단
                break
arr = arr[1:]
print(*arr)


```

```python
# 2차 시도
# 정답
# range 를 활용한 배수 표현
# 20개씩 슬라이싱


N = int(input())            # 스위치 개수
arr = list(map(int,input().split()))
arr = [0] + arr     #인덱스 맞춰주기 위해 값 추가
M = int(input())            # 학생 수
for _ in range(M) :
    sex, num = map(int,input().split())

    # 남학생일 때
    if sex == 1 :
        for i in range(num,N+1,num) :
            arr[i] = 1-arr[i]
    # 여학생일 때
    else :
        arr[num] = 1 - arr[num]
        for i in range(1,N//2+1) :
            left = num-i
            right = num+i
            if left < 1 or right > N : # 범위 벗어나면 중단
                break
            if arr[left] == arr[right] :
                arr[left] = arr[right] = 1 - arr[left]
            else :
                break       #대칭이 아니면 중단

arr = arr[1:]
for x in range(0,len(arr),20) :
    print(*arr[x:x+20])

```
