- 보물섬지도는 직사각형 모양
- 육지(L), 바다(W) 
- 상하좌우로 이웃한 육지로만 이동 가능
- 한 칸 이동하는데 한 시간이 걸림
- 최단거리로 이동하는데 있어 가장 긴 시간이 걸리는 육지 두 곳에 나뉘어 묻혀 있다.
- 최단거리로 가려면 같은 곳을 다시 밟거나, 멀리 돌아가서는 안된다

```python
# 1차 시도 : 실패
# 마지막에 남은 visit[nr][nc]가 항상 최대값이라는 보장이 없음

def bfs(i,j,N,M) :


    # 준비
    visit = [[0] * M for _ in range(N)]
    q = []
    visit[i][j] = 1
    q.append([i,j])


    #탐색
    while q :
        r,c = q.pop(0)
        for dr,dc in ([0,1],[1,0],[0,-1],[-1,0]) :
            nr,nc = r+dr, c+dc
            if 0<= nr < N and 0 <= nc < M and arr[nr][nc] == 'L' and visit[nr][nc] == 0 :
                visit[nr][nc] = visit[r][c] +1
                q.append([nr,nc])
    return visit[nr][nc]



    # for line in visit:
    #     print(*line)
    # print()

# 보물 지도의 세로와 가로 크기
N,M = map(int,input().split())
arr = [input() for _ in range(N)]

ans = 0


for i in range(N) :
    for j in range(M) :
        if arr[i][j] == 'L' :
            cnt = bfs(i,j,N,M)
            if ans < cnt :
                ans = cnt

print(ans)
```

```python
# 2차 시도 : cnt하는 방식으로 했으나 너비우선탐색이라 인접 노드 간것도 다 cnt+=1 되버려서 실패
# 3차 시도(아래) : 시간 초과


def bfs(i,j,N,M) :
    # 준비
    global long
    visit = [[0] * M for _ in range(N)]
    q = []
    visit[i][j] = 1
    q.append([i,j])


    #탐색
    while q :
        r,c = q.pop(0)
        for dr,dc in ([0,1],[1,0],[0,-1],[-1,0]) :  # 우하좌상 확인
            nr,nc = r+dr, c+dc
            if 0<= nr < N and 0 <= nc < M and arr[nr][nc] == 'L' and visit[nr][nc] == 0 :
                q.append([nr,nc])
                visit[nr][nc] = visit[r][c] +1      # 최단거리 기록
                if long < visit[nr][nc] :
                    long = visit[nr][nc]            # 최단거리 중 최장거리기록

    return long



    # for line in visit:
    #     print(*line)
    # print()

# 보물 지도의 세로와 가로 크기
N,M = map(int,input().split())
arr = [input() for _ in range(N)]
long = 0
ans = 0


for i in range(N) :
    for j in range(M) :
        if arr[i][j] == 'L' :
            bfs(i,j,N,M)
            if ans < long-1 :
                ans = long-1

print(ans)
```

```python
# 3차 시도 : 시간초과
# deque를 이용해 pop(0)(시간복잡도 : O(N)) 연산을 없애고
# popleft()(시간복잡도 O(1)0로 바꿈
# max 활용

from collections import deque
def bfs(i,j,N,M) :
    # 준비
    visit = [[0] * M for _ in range(N)]
    q = deque()
    q.append([i,j])
    visit[i][j] = 1
    max_dist = 0        #현재 위치에서의 최장 거리


    #탐색
    while q :
        r,c = q.popleft()
        for dr,dc in ([0,1],[1,0],[0,-1],[-1,0]) :  # 우하좌상 확인
            nr,nc = r+dr, c+dc
            if 0<= nr < N and 0 <= nc < M and arr[nr][nc] == 'L' and visit[nr][nc] == 0 :
                q.append([nr,nc])
                visit[nr][nc] = visit[r][c] +1      # 최단거리 기록
                if max_dist < visit[nr][nc] :
                    max_dist = visit[nr][nc]            # 최단거리 중 최장거리기록

    return max_dist-1   # 처음 시작 위치가 1로 시작하므로 1 빼주기

    # for line in visit:
    #     print(*line)
    # print()

# 보물 지도의 세로와 가로 크기
N,M = map(int,input().split())
arr = [input() for _ in range(N)]
ans = 0


for i in range(N) :
    for j in range(M) :
        if arr[i][j] == 'L' :
            ans = max(ans,bfs(i,j,N,M))
print(ans)

```

```python
#응용, 가장 먼 지점 찾기(BFS 횟수 감소)

import sys
from collections import deque

sys.stdin = open("test.txt")

def bfs(start_i, start_j, N, M):
    # BFS를 위한 큐와 방문 체크 리스트 준비
    visit = [[0] * M for _ in range(N)]
    q = deque()
    q.append((start_i, start_j))
    visit[start_i][start_j] = 1
    
    max_dist = 0
    farthest_point = (start_i, start_j)
    
    # BFS 탐색 시작
    while q:
        r, c = q.popleft()
        for dr, dc in ([0, 1], [1, 0], [0, -1], [-1, 0]):  # 상하좌우 탐색
            nr, nc = r + dr, c + dc
            if 0 <= nr < N and 0 <= nc < M and arr[nr][nc] == 'L' and visit[nr][nc] == 0:
                q.append((nr, nc))
                visit[nr][nc] = visit[r][c] + 1  # 최단 거리 기록
                # 최장 거리 및 가장 먼 지점 업데이트
                if visit[nr][nc] > max_dist:
                    max_dist = visit[nr][nc]
                    farthest_point = (nr, nc)
    
    return max_dist - 1, farthest_point  # 거리에서 1을 빼줌


# 보물 지도의 세로와 가로 크기 입력 받기
N, M = map(int, input().split())
arr = [input().strip() for _ in range(N)]
max_distance = 0  # 모든 육지에서의 최대 거리

# 첫 번째 BFS: 가장 먼 지점 찾기
for i in range(N):
    for j in range(M):
        if arr[i][j] == 'L':
            # 가장 먼 지점과 그곳까지의 거리 반환
            _, farthest_point = bfs(i, j, N, M)
            break  # 한 번만 수행하고 나감

# 두 번째 BFS: 가장 먼 지점에서 다시 BFS 수행하여 최대 거리 찾기
if farthest_point:  # 가장 먼 지점이 있는 경우
    max_distance, _ = bfs(farthest_point[0], farthest_point[1], N, M)

print(max_distance)
```
