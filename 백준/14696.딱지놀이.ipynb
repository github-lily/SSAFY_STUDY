- 두 어린이는 여러장의 딱지를 가지고 있고, 매 라운드마다 하나씩 낸다
- 딱지에는 별(★,4), 동그라미(●,3), 네모(■,2), 세모(▲,1), 네 가지 모양 중 하나 이상의 모양이 표시
- 라운드별로 딱지놀이의 결과를 출력

### 순위결정
1. 별의 개수가 다르다면 별 많은쪽이 이긴다
2. 동그라미
3. 네모
4. 세모
5. 개수가 모두 같으면 무승부(D)

```python
# 1차 시도 : 실패
# 맨 앞글자는 모양의 개수이기 떄문에 같이 정렬되면 안됨
import sys

sys.stdin = open("test.txt")

def winner(A,B) :

    for i in range(1, max(len(A),len(B))) :  #첫번째 값 제외

        # 비교 대상이 범위 밖이라면, 그동안 판가름이 안났단 뜻이므로 더 긴게 답
        if i >= len(A) :
            return 'B'
        if i >= len(B) :
            return 'A'


        if A[i] > B[i] :
            return 'A'
        elif A[i] == B[i] :
            continue
            # 둘 다 값이 같으면 다음 자리에서 판단해야하니 계속 진행
        else :
            return 'B'
    return 'D'





N = int(input())    #라운드 총 수
for _ in range(N) :
    A = list(map(int,input().split()))   # A[0] = a : 그 라운드에서 A가 가진 모양의 총 개수
    B = list(map(int,input().split()))   # B[0] = b : 그 라운드에서 B가 가진 모양의 총 개수

    # 우선순위가 높은 수가 앞에 오도록 역정렬(내림차순정렬)
    A.sort(reverse=True)
    B.sort(reverse=True)

    result = winner(A,B)

    print(result)
```
