```python
# 1차 시도 : 시간초과


def dfs(sr,sc, n,number) :      #sr,sc : 시작점
    global ans

    if n == 7 :
        num_set.add(number)
        ans += 1
        return


    dr = [0,1,0,-1]
    dc = [1,0,-1,0]


    for i in range(4) :
        for j in range(N) :
            nr = sr + dr[i] * j
            nc = sc + dc[i] * j
            if 0 <= nr < N and 0 <= nc < N :
                dfs(nr,nc,n+1, number + str(arr[nr][nc]))

    



T = int(input())
for tc in range(1,T+1) :
    N = 4
    arr = [list(map(int,input().split())) for _ in range(N)]

    num_set = set()
    number = ''
    ans = 0


    for r in range(N) :
        for c in range(N) :
            dfs(r,c,0,number+str(arr[r][c]))

    print(f'#{tc} {ans}')
```

```python
# 2차 시도 : 통과


def dfs(sr,sc, n,number) :      #sr,sc : 시작점

    if n == 7 :
        num_set.add(number)
        return


    dr = [0,1,0,-1]
    dc = [1,0,-1,0]


    for i in range(4) :
        nr = sr + dr[i]
        nc = sc + dc[i]
        if 0 <= nr < N and 0 <= nc < N :
            dfs(nr,nc,n+1, number + str(arr[nr][nc]))

    



T = int(input())
for tc in range(1,T+1) :
    N = 4
    arr = [list(map(int,input().split())) for _ in range(N)]

    num_set = set()
    number = ''



    for r in range(N) :
        for c in range(N) :
            dfs(r,c,1,number+str(arr[r][c]))

    print(f'#{tc} {len(num_set)}')
```
