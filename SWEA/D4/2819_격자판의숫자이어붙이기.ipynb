- 4x4 크기의 격자판, 0~9 사이의 숫자
- 임의의 위치에서 시작
- 동서남북 네 방향으로 인접한 격자로 총 6번 이동
- 각 칸에 적혀있는 숫자를 차례대로 이어붙여 7자리 숫자를 만든다
- 재방문 가능 : visit 필요 없음
- 0으로 시작하는 수를 만들 수 있음
- 만들 수 있는 서로 다른 일곱 자리 수들의 개수는? : set
- 제한시간 8초!

```python
# 1차 시도 : 시간초과


def dfs(sr,sc, n,number) :      #sr,sc : 시작점
    global ans

    if n == 7 :
        num_set.add(number)
        ans += 1
        return


    dr = [0,1,0,-1]
    dc = [1,0,-1,0]


    for i in range(4) :
        for j in range(N) :
            nr = sr + dr[i] * j
            nc = sc + dc[i] * j
            if 0 <= nr < N and 0 <= nc < N :
                dfs(nr,nc,n+1, number + str(arr[nr][nc]))

    



T = int(input())
for tc in range(1,T+1) :
    N = 4
    arr = [list(map(int,input().split())) for _ in range(N)]

    num_set = set()
    number = ''
    ans = 0


    for r in range(N) :
        for c in range(N) :
            dfs(r,c,0,number+str(arr[r][c]))

    print(f'#{tc} {ans}')
```

```python
# 2차 시도 : 통과


def dfs(sr,sc, n,number) :      #sr,sc : 시작점

    if n == 7 :
        num_set.add(number)
        return


    dr = [0,1,0,-1]
    dc = [1,0,-1,0]


    for i in range(4) :
        nr = sr + dr[i]
        nc = sc + dc[i]
        if 0 <= nr < N and 0 <= nc < N :
            dfs(nr,nc,n+1, number + str(arr[nr][nc]))

    



T = int(input())
for tc in range(1,T+1) :
    N = 4
    arr = [list(map(int,input().split())) for _ in range(N)]

    num_set = set()
    number = ''



    for r in range(N) :
        for c in range(N) :
            dfs(r,c,1,number+str(arr[r][c]))

    print(f'#{tc} {len(num_set)}')
```

```python
# 문어박사님 코드 복기 : 성공

def dfs(n,num,cr,cc) :
    if n == 7 :
        num_set.add(num)
        return
    
    dr = [0,1,0,-1]
    dc = [1,0,-1,0]

    for i in range(4) :
        nr = cr+dr[i]
        nc = cc+dc[i]
        if 0 <= nr < 4 and 0 <= nc < 4 :
            dfs(n+1, num*10 + arr[nr][nc], nr, nc)
            
    

T = int(input())
for tc in range(1,T+1) :
    arr = [list(map(int,input().split())) for i in range(4)]

    num_set = set()

    for r in range(4) :
        for c in range(4) :
            dfs(1,arr[r][c],r,c)
            # arr[r][c] 에 있는 값을 넣고 시작, 그 뒤는 현재 위치

    print(f'#{tc} {len(num_set)}')

```
