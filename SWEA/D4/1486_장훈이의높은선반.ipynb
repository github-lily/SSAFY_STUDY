```python
# 9.14 복기 : 성공

def dfs(n,sm) :
    global ans
     
    # 가지치기
    if sm-B > ans :
        return
 
    if n == N :
        if sm >= B :
            ans = min(sm-B, ans)
        return
 
    dfs(n+1,sm+lst[n])
    dfs(n+1,sm)
 
 
T = int(input())
for tc in range(1,T+1) :
    N, B = map(int,input().split())
    lst = list(map(int,input().split()))
    ans = 9999999999999999999999
 
    dfs(0,0)
 
    print(f'#{tc} {ans}')
```

```python
# 1차 시도 : 성공
# 하지만 실행시간이 많이 느림


T = int(input())
for tc in range(1,T+1) :
    N, B = map(int,input().split())
    arr = list(map(int,input().split()))
    min_high = 99999999999999999999

    for subset in range(1 << N):
        tmp = []
        for idx in range(N):
            if subset & (1 << idx):
                tmp += [arr[idx]]
            sum_val = sum(tmp)
            if min_high < sum_val :
                break

        if sum_val >= B :
            if min_high > sum_val :
                min_high = sum_val

    print(f'#{tc} {min_high-B}')
```

```python
# 현준오빠 풀이방식. 수정했지만 계속 오답이 나옴... gpt도 틀림..
# 내 생각엔 01111 이 답인데 그 경우의 수가 안나오기 때문인 듯


def backtrack(bitmask, cur_sum):
    global min_val

    # 총 1이 N개 채워지면
    if bitmask == (1 << N) - 1:  # 10000 -> 4개의 원소보다 크므로 개념이 성립하지 않음
        if cur_sum >= B:
            diff = cur_sum - B
            if min_val > diff:
                min_val = diff
        return


    # 0000 ~ 1111
    for lsh in range(N):  # lsh(left shift)
        if not bitmask & (1 << lsh):  # lsh를 인덱스로 하는 위치에 1이 표시되어 있는가
            backtrack(bitmask | (1 << lsh), cur_sum + arr[lsh])  # | or연산자

# Main part of the code
T = int(input())
for tc in range(1, T + 1):
    N, B = map(int, input().split())
    arr = list(map(int, input().split()))

    min_val = float('inf')  # 최소값을 무한대로 초기화

    backtrack(0, 0)

    # 만약 min_val이 여전히 무한대라면, B 이상의 탑이 없었음을 의미
    if min_val == float('inf'):
        min_val = 0

    print(min_val)
```
