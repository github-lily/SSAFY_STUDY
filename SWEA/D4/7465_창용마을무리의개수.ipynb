```python

'''
방법1
- 부모 인덱스가 자기 자신을 가리키고 있는 것이 루트다
- 개수 카운트

방법2
- 처음엔 각각 개별 집합으로 존재 (N)개
- 다른 노드와 연결될 때마다 1씩 감소시키기
'''





T = int(input())
for tc in range(1,T+1) :
    V,E = map(int,input().split())

    # make_set
    p = [i for i in range(V+1)]     # 1번부터 시작하므로 +1

    def find_set(x) :
        if x != p[x] :
            p[x] = find_set(p[x])
        return p[x]

    ans = V
    for _ in range(E) :
        u,v = map(int,input().split())

        a = find_set(u)
        b = find_set(v)
        if a == b :
            continue        # 조상이 같단 뜻
        p[a] = b
        ans -= 1            # 그룹이 갈라짐(방법2)

    print(f'#{tc} {ans}')
```

```python
# 교수님 코드

T = int(input())
 
for tc in range(1, T + 1):
    V, E = map(int, input().split())
    # -----------------------------------------
    # 정점 번호: 1 ~ V
    p = [i for i in range(V + 1)]
    # x가 속한 집합(트리)의 대표자(루트)를 반환
    def find_set(x):
        if x == p[x]:
            return x
        else:
            p[x] = find_set(p[x])
            return p[x]
    # -----------------------------------------
    ans = V
    for _ in range(E):
        u, v = map(int, input().split())
        a = find_set(u)
        b = find_set(v)
        if a == b:
            continue
        p[a] = b
        ans -= 1
 
    print(f'#{tc} {ans}')
```
