# 1232. 사칙연산

- 정점에 연산자가 있으면 해당 연산자의 왼쪽 서브트리 결과와 오른쪽 서브트리 결과에 해당 연산자를 적용한다

- 정수이면 정점번호와 양의 정수
- 연산자이면 정점 번호, 연산자, 왼쪽 자식, 오른쪽 자식
- 값 나눠서 받아 입력
- 아래 노드부터 자식 노드가 사칙연산이 아니라면으로 찾아서 계산


```python
# 사칙연산 계산할 때 값 나눠서 하는거 주의
# 인덱스 값 항상 생각하기
# int로 형변환 잊지말기
# left,right 리스트
# 실행시간은 제일 빠름

for tc in range(1,11) :
    N = int(input())

    tree = [0] * (N+1)
    left = [0] * (N+1)
    right = [0] * (N+1)

    # tree에 값 저장
    for _ in range(N) :
        temp = list(input().split())
        if len(temp) == 2 :
            tree[int(temp[0])] = int(temp[1])
        else :
            tree[int(temp[0])] = temp[1]
            left[int(temp[0])] = int(temp[2])
            right[int(temp[0])] = int(temp[3])

    cal = ['+','-','*','/']

    for v in range(N,0,-1) :
        if tree[v] in cal : #사칙연산이라면
            if left[v] != 0 and right[v] != 0 :
                if tree[v] == '+' :
                    tree[v] = tree[left[v]] + tree[right[v]]
                elif tree[v] == '-' :
                    tree[v] = tree[left[v]] - tree[right[v]]
                elif tree[v] == '*' :
                    tree[v] = tree[left[v]] * tree[right[v]]
                elif tree[v] == '/' :
                    tree[v] = tree[left[v]] // tree[right[v]]


    print(f'#{tc} {tree[1]}')


```

```python
for tc in range(1, 11):
    N = int(input())
    L = [0] * (N + 1)
    R = [0] * (N + 1)
    data = [0] * (N + 1)
 
    for _ in range(N):
        tmp = input().split()
        idx = int(tmp[0])
        if len(tmp) == 4:
            data[idx] = tmp[1]
            L[idx] = int(tmp[2])
            R[idx] = int(tmp[3])
        else:
            data[idx] = int(tmp[1])
 
    def calc(v):
        if L[v] == 0:
            return data[v]
 
        l = calc(L[v])
        r = calc(R[v])
 
        if data[v] == '+': ret = l + r
        elif data[v] == '-': ret = l - r
        elif data[v] == '*': ret = l * r
        else: ret = l // r
        return ret
 
    print(f'#{tc} {calc(1)}')
```
