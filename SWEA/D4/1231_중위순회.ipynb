```python
def tree(node) :
    global text
    if node == 0 :
        return 0

    tree(left[node])
    # 중위순회
    # text에 node 인덱스의 문자 저장
    text += char[node]
    tree(right[node])




for tc in range(1,11) :
    N = int(input())    # 트리의 정점 개수
    E = N-1     #간선의 수

    char = [0] * (N + 1)
    left = [0] * (N + 1)
    right = [0] * (N + 1)
    text = ''


    for _ in range(N) :
        # 정점 번호, 정점 문자, 왼쪽 자식, 오른쪽 자식
        # 스트링이 껴있어서 int로 한번에 받을 수 없음
        # 숫자는 인트로 변환해주는거 주의하기
        temp = list(input().split())
        char[int(temp[0])] = temp[1]

        # 값이 없는 경우도 있으므로 길이로 나누기
        if len(temp) >= 3 :
            left[int(temp[0])] = int(temp[2])
            if len(temp) >= 4:
                right[int(temp[0])] = int(temp[3])



    tree(1)
    print(f'#{tc} {text}')
```
