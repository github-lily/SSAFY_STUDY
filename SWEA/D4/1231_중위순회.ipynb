```python
# 9월 1일 복기
# 재귀함수 종료조건 신경쓰기


import sys
sys.stdin = open("test.txt")

'''
- 중위순회하면서 나오는 단어를 출력하라
'''

'''
1. 트리에 단어를 담는다
2. 중위순회한다
'''


def inorder(node) :
    global text
    # node 값이 0 이면 중단!! 잊지말기
    if node > N  or node == 0:
        return

    inorder(left[node])
    text += tree[node]
    inorder(right[node])

for tc in range(1,11) :
    N = int(input())    # 정점의 개수

    tree = [0] * (N+1)
    left = [0] * (N+1)
    right = [0] * (N+1)
    text = ""

    # 입력 정보 저장
    for _ in range(N) :
        temp = list(input().split())
        if len(temp) < 3 :      # 단말 노드일 경우
            tree[int(temp[0])] = temp[1]

        elif len(temp) == 3 :
            tree[int(temp[0])] = temp[1]
            left[int(temp[0])] = int(temp[2])

        else :
            tree[int(temp[0])] = temp[1]
            left[int(temp[0])] = int(temp[2])
            right[int(temp[0])] = int(temp[3])

    inorder(1)
    print(f'#{tc} {text}')


```
