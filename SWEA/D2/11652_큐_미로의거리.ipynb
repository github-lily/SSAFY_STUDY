```python
import sys
sys.stdin = open("test.txt", "r")




def bfs(sti,stj,lst,N) :
    visited = [0] * (N+1)
    q = []
    q.append([sti,stj])
    visited[sti][stj] = 1

    while q :
        vi, vj = q.pop(0)
        if lst[vi][vj] == 3:        #종료지점이면 끝
            return visited[vi][vj]
        else :
            for di, dj in [[0,1],[1,0],[0,-1],[-1,0]] :     #주변 탐색
                wi = vi+di
                wj = vj+dj
                if 0 <= wi < N and 0 <= wj < N and lst[wi][wj] != 1 and visited[wi][wj] == 0 :
                    q.append([wi,wj])
                    visited[wi][wj] = visited[vi][vj] +1



def find_s(lst,N) :             #maze는 전역변수라서 따로 lst로 정해주지 않아도 바로 가져다가 사용할 수 있음
    for i in range(N) :
        for j in range(N) :
            if lst[i][j] == 2 :
                return [i,j]


T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    maze = [list(map(int, input())) for _ in range(N)]



    sti,stj = find_s(maze,N)
    result = bfs(sti,stj,maze,N)


    print(f'#{tc} {result}')
```
