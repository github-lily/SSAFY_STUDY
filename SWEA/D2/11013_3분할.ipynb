# 03_배열1_3분할
- 3개의 연속적인 구간으로 나뉨
- 각 구간의 총합 중 최대값과 최소값의 차이가
- 최소가 되는 경우의 차이값

```python
T = int(input())
for tc in range(1, T+1):
    N = int(input())
    lst = list(map(int, input().split()))

    ans = 9999999999
    for i in range(0,N-2) :
        for j in range(i+1,N-1) :       #i다음부터 시작인거 주의!!
            sum_1 = sum(lst[:i+1])
            sum_2 = sum(lst[i+1:j+1])
            sum_3 = sum(lst[j+1:])

            sum_max = max(sum_1,sum_2,sum_3)
            sum_min = min(sum_1,sum_2,sum_3)

            diff = sum_max - sum_min
            if ans > diff :
                ans = diff





    print(f'#{tc} {ans}')
```

```python
# 함수 연습
# 함수로 풀어보기
# 초기화 위치 항상 주의하기



def find_sum(s,e) :
    sum_val = 0
    for k in lst[s:e] :
        sum_val += k
    return sum_val

def find_max() :
    max_val = -9999999      #음수도 있으니 낮게 잡기
    for x in sum_lst :
        if max_val < x :
            max_val = x
    return max_val

def find_min() :
    min_val = 99999999
    for y in sum_lst :
        if min_val > y :
            min_val = y
    return min_val


T = int(input())
for tc in range(1, T+1):
    N = int(input())
    lst = list(map(int, input().split()))

    ans = 9999999999
    for i in range(0,N-2) :
        for j in range(i+1,N-1) :
            sum_lst = []
            sum_lst.append(find_sum(0,i+1))
            sum_lst.append(find_sum(i+1, j+1))
            sum_lst.append(find_sum(j+1, N))

            diff = find_max() - find_min()

            if ans > diff :
                ans = diff





    print(f'#{tc} {ans}')
```
