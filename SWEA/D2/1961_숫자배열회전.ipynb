# 1961 숫자배열회전
- NxN 행렬을 90도 180도 270도 회전한 모양을 출력
- 3 <= N <= 7
- k열 -> k 행 -> n-1-k열 -> n-1-k행
- arr[][k] -> arr[k][] -> arr[][n-1-k] -> arr[n-1-k][]
- 한 줄에 90도 180도 270도의 결과를 출력


```python
# 1차 시도
# 굳이 for를 3개 사용할 필요가 없음
# 행과 열을 함께 바꿔줄 수 있음

T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]

    #90도 회전
    for i in range(N) :
        for j in range(N) :
            for k in range(N) :
                arr[i][k] = arr[k][i]
    
    print(arr)
```

```python
# 2차 시도

T = int(input())
for tc in range(1 , T +1) :
    N = int(input())
    arr = [list(map(int ,input().split())) for _ in range(N)]
    new_arr = [[0] * N for _ in range(N)]


    # 90도 회전
    for i in range(N) :
        for j in range(N) :
            new_arr[j][N-1-i] = arr[i][j]       # 90도 회전


    print(*new_arr)

    # 180도 270도를 한 줄에 어떻게 출력해야하는지 모르겠음
```

```python
# 3차 시도
# arr을 그대로 덮어쓰면 수정된 값이 수정되는 오류가 생김
# 고로 새로운 리스트 만들어줘야함

T = int(input())
for tc in range(1 , T +1) :
    N = int(input())
    arr = [list(map(int ,input().split())) for _ in range(N)]
    
    

    arr_90 = [[0] * N for _ in range(N)]
    arr_180 = [[0] * N for _ in range(N)]
    arr_270 = [[0] * N for _ in range(N)]


    for i in range(N) :
        for j in range(N) :
            arr_90[j][N-1-i] = arr[i][j]       # 90도 회전
            arr_180[j][N-1-i] = arr_90[i][j]       # 180도 회전
            arr_270[j][N-1-i] = arr_180[i][j]       # 270도 회전
       



    print(*arr_90)
    print(*arr_180)
    print(*arr_270)

    # 한 for 문 안에서 돌리니 값이 제대로 안나옴
    # arr 90이 모두 채워지지 않은 상태에서 180으로 이동시키니 오류가 나옴
```

```python
# 4차 시도
# for문 각각으로 바꿈

T = int(input())
for tc in range(1 , T +1) :
    N = int(input())
    arr = [list(map(int ,input().split())) for _ in range(N)]
    
    

    # 90도 회전
    arr_90 = [[0] * N for _ in range(N)]
    for i in range(N) :
        for j in range(N) :
            arr_90[j][N-1-i] = arr[i][j]       

    # 180도 회전
    arr_180 = [[0] * N for _ in range(N)]
    for i in range(N) :
        for j in range(N) :
            arr_180[j][N-1-i] = arr_90[i][j]       

    # 270도 회전
    arr_270 = [[0] * N for _ in range(N)]
    for i in range(N) :
        for j in range(N) :        
            arr_270[j][N-1-i] = arr_180[i][j]       



    for i in range(3) :
        print(*map(str, arr_90[i], arr_180[i], arr_270[i]))
        # map이 출력됨
```

```python
# 5차 시도
# 성공
# join 사용해서 map 출력되는걸 바꿈

T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    # 90도 회전
    arr_90 = [[0] * N for _ in range(N)]
    for i in range(N):
        for j in range(N):
            arr_90[j][N - 1 - i] = arr[i][j]

    # 180도 회전
    arr_180 = [[0] * N for _ in range(N)]
    for i in range(N):
        for j in range(N):
            arr_180[j][N - 1 - i] = arr_90[i][j]


    # 270도 회전
    arr_270 = [[0] * N for _ in range(N)]
    for i in range(N):
        for j in range(N):
            arr_270[j][N - 1 - i] = arr_180[i][j]


    print(f'#{tc}')
    for i in range(N):  #N개만큼 줄이 출력해야 하므로 N
        line_90 = ''.join(map(str,arr_90[i]))
        line_180 = ''.join(map(str,arr_180[i]))
        line_270 = ''.join(map(str,arr_270[i]))
        print(line_90, line_180, line_270)
```

```python
# 6차 시도
# 성공
# 기본 배열 arr 에서 90도 180도 270도 행렬의 변화를 확인해서 값을 바꿈

T = int(input())
for tc in range(1 , T +1) :
    N = int(input())
    arr = [list(map(int ,input().split())) for _ in range(N)]
    
    

    arr_90 = [[0] * N for _ in range(N)]
    arr_180 = [[0] * N for _ in range(N)]
    arr_270 = [[0] * N for _ in range(N)]


    for i in range(N) :
        for j in range(N) :
            arr_90[j][N-1-i] = arr[i][j]       # 90도 회전
            arr_180[N-1-i][N-1-j] = arr[i][j]    # 180도 회전
            arr_270[N-1-j][i] = arr[i][j]       # 270도 회전
       


    print(f'#{tc}')
    for i in range(N):  #N개만큼 줄이 출력해야 하므로 N
        line_90 = ''.join(map(str,arr_90[i]))
        line_180 = ''.join(map(str,arr_180[i]))
        line_270 = ''.join(map(str,arr_270[i]))
        print(line_90, line_180, line_270)
```

```python
# 전준표_1226826 님 코드

for tc in range(int(input())):
    n = int(input())
    arr = [list(map(int, input().split())) for _ in range(n)]
     
    ans = [[] * n for _ in range(n)]
    for _ in range(3):       #90도, 180도, 270도 3개니까 N = 3
        new_arr = [[0] * n for _ in range(n)]
        for x in range(n):
            new_line_i = []
            for y in range(n - 1, -1, -1):
                new_line_i.append(arr[y][x])
            new_arr[x] = new_line_i
            ans[x].append(''.join(map(str, new_line_i)))
        arr = new_arr
     
    print(f'#{tc + 1}')
    for i in range(n):
        print(' '.join(ans[i]))
```

```python
# 황다빈_1248591 님 코드

def rotate_90(N, arr):
    # 90도 회전
    new_arr = [[0] * N for _ in range(N)]
    for i in range(N):
        for j in range(N):
            new_arr[j][N-1-i] = arr[i][j]
    return new_arr
 
def rotate_180(N, arr):
    # 180도 회전
    return rotate_90(N, rotate_90(N, arr))
 
def rotate_270(N, arr):
    # 270도 회전
    return rotate_90(N, rotate_180(N, arr))
 
T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]
 
    result_90 = rotate_90(N, arr)
    result_180 = rotate_180(N, arr)
    result_270 = rotate_270(N, arr)
 
    print(f'#{tc}')
    for i in range(N):
        # 각 줄의 3개의 배열을 이어붙여서 출력
        line_90 = ''.join(map(str, result_90[i]))
        line_180 = ''.join(map(str, result_180[i]))
        line_270 = ''.join(map(str, result_270[i]))
        print(line_90, line_180, line_270)
```

```python
# VIN 님 코드 (실행시간 제일 짧음)
T = int(input())
result = list()
for test_case in range(1, T + 1):
    N = int(input())
    if 3 <= N <= 7:
        p = [list(map(str, input().split())) for _ in range(N)]
        p90 = list(zip(*p[::-1]))
        p180 = list(zip(*p90[::-1]))
        p270 = list(zip(*p180[::-1]))
        st = '#{}'.format(test_case)
        for i in range(len(p)):
            temp = ''.join(p90[i]) + ' ' + ''.join(p180[i]) + ' ' + ''.join(p270[i])
            st = st + '\n' + temp
        result.append(st)
 
for test_case in range(1, T + 1):
    print(result[test_case - 1])
```
