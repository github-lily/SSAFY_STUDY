# 1979 어디에 단어가 들어갈 수 있을까
- NxN 크기의 단어 퍼즐을 만드려고 한다.
- 길이 K인 단어가 들어갈 수 있는 자리의 수를 출력하라

- 가로 세로 모두 고려해야 함
- 행우선탐색 열우선탐색 활용
- K개 연속된 값의 최대 개수 구하기 활용

- 흰색 1, 검은색 0

```python
# 1차 시도

T = int(input())
for tc in range(1,T+1) :
    N,K = map(int,input().split())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0
    for r in range(N) :
        cnt = 0     #새로운 행을 탐색할 때마다 새롭게 카운트
        for c in range(N) :
            if arr[r][c] == 1 :
                cnt += 1        #연속되지 않은 경우에도 1이면 무조건 +=1이 실행되버림 그래서 오류
        if cnt == K :   
            ans += 1

    for c in range(N) :
        cnt = 0
        for c in range(N) :
            if arr[r][c] == 1 :
                cnt += 1
        if cnt == K :
            ans += 1



    print(f'#{tc} {ans}')
```

```python
# 2차 시도 : 성공
# 연속된 값일 때 cnt 증가하도록 설정
# 카운트 종료 들여쓰기 잘 보기
# 실행시간 122ms

T = int(input())
for tc in range(1,T+1) :
    N,K = map(int,input().split())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0     #최종 답
    #열우선탐색
    for r in range(N) :
        cnt = 1     #연속된 값일 때 증가하므로 기본값1
        for c in range(1, N) :
            if arr[r][c-1] == arr[r][c] == 1:
               cnt += 1
            else :
                if cnt == K :
                    ans +=1
                cnt = 1     #연속이 끊기면 리셋
        if cnt == K :
            ans +=1

    #행우선탐색
    for j in range(N) :
        cnt = 1
        for i in range(1, N) :
            if arr[i-1][j] == arr[i][j] == 1:
               cnt += 1
            else :
                if cnt == K :
                    ans +=1
                cnt = 1
        if cnt == K :
            ans +=1


    print(f'#{tc} {ans}')
```

```python
# 함수버전
# 실행시간 185ms
# 코드가 간단해서 오히려 함수쓰는게 더 오래걸림


def row() :
    global ans
    for j in range(N) :
        cnt = 1
        for i in range(1, N) :
            if arr[i-1][j] == arr[i][j] == 1:
               cnt += 1
            else :
                if cnt == K :
                    ans +=1
                cnt = 1
        if cnt == K :
            ans +=1
    return ans

def column() :
    global ans
    for r in range(N) :
        cnt = 1
        for c in range(1, N) :
            if arr[r][c-1] == arr[r][c] == 1:
               cnt += 1
            else :
                if cnt == K :
                    ans +=1
                cnt = 1
        if cnt == K :
            ans +=1
    return ans

T = int(input())
for tc in range(1,T+1) :
    N,K = map(int,input().split())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0

    row()
    column()




    print(f'#{tc} {ans}')
```
