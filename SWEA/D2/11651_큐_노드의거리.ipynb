```python
#너비우선탐색 함수 만들기

def bfs(s,g,lst) :
    #준비
    #visit 만들기
    visited = [0] * 51      #정점의 최대 개수 50 +1 (1번부터 시작하니까)
    #큐 만들기
    q = []
    #인큐
    q.append(s)
    #visit 표시하기
    visited[s] = 1

    #탐색
    while q :
        # 디큐
        v = q.pop(0)

        if v == g :
            return visited[v] -1

        else :
        #인접 정점 찾기
            for w in lst[v] :
            #만약 큐 안에 없다면
                if w not in q :
            #인큐
                    q.append(w)
            #visit에 표시
                    visited[w] = visited[v]+1
    else :
        return 0






T = int(input())
for tc in range(1, T + 1):
    V, E = map(int, input().split())
    arr = [list(map(int, input().split())) for _ in range(E)]
    S, G = map(int,input().split())

    match = [[] for _ in range(V+1)]

    #인접 정점 리스트 만들기
    for i in range(E) :
        v1,v2 = arr[i][0], arr[i][1]
        match[v1].append(v2)
        match[v2].append(v1)




    print(f'#{tc} {bfs(S,G,match)}')

    #if w not in q : 조건을 이렇게 걸면 pop으로 나간애들도 다시 들어올 수 있음
```
