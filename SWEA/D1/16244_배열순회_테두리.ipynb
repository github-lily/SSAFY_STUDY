```python
T = int(input())
for tc in range(1,T+1) :
    r1, c1, N = map(int,input().split())
    arr = [[0] * 10 for _ in range(10)]

    arr[r1][c1] = 1
    dr = [0,1,0,-1]
    dc = [1,0,-1,0]


    for i in range(4) :
        for j in range(1,N) :
            nr = r1 + dr[i]
            nc = c1 + dc[i]
            arr[nr][nc] = 1
            r1 = nr
            c1 = nc

    print(f'#{tc}')
    for line in arr :
        print(*line)
```

```python

'''
초기값이 0으로 채워진 10x10 2차 배열
정사각형 모양의 좌상단 모서리와 크기가 주어지면
테두리 부분을 1로 저장한다
'''

# 9.8 복기
# 각 구간으로 나눈 뒤 구간의 처음과 끝을 함수에 넣어 마크하기

def change(rs,cs,re,ce) :
    for row in range(rs,re+1) :
        for col in range(cs,ce+1) :
            arr[row][col] = 1


T = int(input())
for tc in range(1,T+1) :
    r1,c1,w = map(int,input().split())
    arr = [[0]*10 for _ in range(10)]

    change(r1,c1,r1,c1+w-1)
    change(r1,c1+w-1,r1+w-1,c1+w-1)
    change(r1+w-1,c1,r1+w-1,c1+w-1)
    change(r1,c1,r1+w-1,c1)

    print(f'#{tc}')
    for line in arr :
        print(*line)
```

```python

T = int(input())
for tc in range(1,T+1) :
    r1,c1,M = map(int,input().split())

    arr = [[0]*10 for _ in range(10)]
    N = len(arr)


    # 위
    for c in range(c1,c1+M) :
        if c >= N : break
        arr[r1][c] = 1

    # 왼쪽
    for r in range(r1,r1+M) :
        if r >= N: break
        arr[r][c1]  = 1

    # 아래
    for c in range(c1,c1+M) :
        if c >= N and r1+M-1 >=M: break
        arr[r1+M-1][c] = 1

    # 오른쪽
    for r in range(r1,r1+M) :
        if r >= N and c1+M-1 >=M: break
        arr[r][c1+M-1] = 1


    print(f'#{tc}')
    for line in arr :
        print(*line)
```
