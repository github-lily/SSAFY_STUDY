```python
# 백트레킹 + 가지치기 : 204ms
def dfs(n,sm) :
    global ans

    if sm >= ans :
        return
    
    if n > 12 :
        ans = min(ans, sm)
        return
    
    dfs(n+1, sm+plan[n]*day)        # 일일 이용권
    dfs(n+1, sm+mon)                # 월간 이용권
    dfs(n+3, sm+mon3)               # 3개월 이용권
    dfs(n+12, sm+year)              # 연간 이용권
    



T = int(input())
for tc in range(1,T+1) :
    day, mon, mon3, year = map(int, input().split())
    plan = [0] + list(map(int, input().split()))        # 월과 인덱스 맞추기
    ans = 9999999999999999999       #지출의 최솟값

    dfs(1,0)        # 1월부터 시작


    print(f'#{tc} {ans}')
```

```python

# 그리디
# 각 월마다의 최솟값이 최종 최솟값일 것이다!
# 각 월마다 최솟값을 비교해봄
# 한 해가 아닌 몇년(장기)로 구할 때 백트레킹은 시간초과 위험 그럴땐 이렇게!



T = int(input())
for tc in range(1,T+1) :
    day, mon, mon3, year = map(int, input().split())
    plan = [0] + list(map(int, input().split()))        # 월과 인덱스 맞추기
    ans = 9999999999999999999       # 지출의 최솟값
    D = [0] * 13                    # 각 월의 최솟값을 담을 리스트

    for i in range(1,13) :
        mn = D[i-1] + day * plan[i]
        mn = min(mn, D[i-1] + mon)       # 이전달까지 최소비용 + 월 이용권 금액과 일간권 비교
        if i>= 3 :                       # 3개월 이용권을 쓰려면 3월이 지나야 함
            mn = min(mn, D[i-3] + mon3)  # 지금까지의 최소값과 3개월치 요금 비교
        if i>= 12 :                      # 한 해가 지났다면 연간 이용권과 비교
            mn = min(mn, D[i-12] + year)
        D[i] = mn

    ans = D[12]

    print(f'#{tc} {ans}')
```
