- 맨 왼쪽 위(0,0)에서 오른쪽 아래 (N-1,N-,1)까지 이동할 때
합계가 최소인 루트의 합계는?

```python
# 1차시도
# min_sum의 reset 위치를 찾지 못해서 실패

def KFC(r,c) :
    global ans, min_sum

    min_sum += arr[r][c]

    # 배열의 끝에 도착하면 종료
    if r == N-1 and c == N-1 :
        if ans >= min_sum :
            ans = min_sum
        return ans

    # 우측 열 탐색
    if c+1 < N :
        KFC(r,c+1)

    # 아래쪽 행 탐색
    if r+1 <N :
        KFC(r+1,c)




T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]
    ans = 999         # 최종 답
    min_sum = 0     # 케이스의 합


    KFC(0,0)

    print(f'#{tc} {ans}')
```
