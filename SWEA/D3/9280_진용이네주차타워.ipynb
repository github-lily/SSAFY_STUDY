```python
# 1차 시도
# 틀렸지만 인덱스 오류를 여러번 수정하면서 공부가 됐다!


import sys

sys.stdin = open("test.txt")


'''
- 주차 가능한 공간 중 번호가 가장 작은 주차공간에 주차함
- 주차요금 : 차량의 무게 * 주차 공간마다 따로 책정된 단위 무게당 금액
- 진용이가 하루 벌어들일 총 수입은?
'''


T = int(input())
for tc in range(1,T+1) :
    total = 0           #최종 답
    # 준비과정
    N,M = map(int,input().split())      # N : 1~N개의 주차 공간,주차 예정 차량 M대
    weight_fee = [0] * N                 # 인덱스 주의하기! 주차공간은 1번부터 시작, 인덱스는 0번부터
    for n in range(N) :
        weight_fee[n] = int(input())     # 단위 무게당 요금
    car_weight = [0] * M
    for m in range(M) :
        car_weight[m] = int(input())



    visit = [0] * N     # 차량의 주차 여부 기록
    for j in range(2* M) :
        flow = int(input())      # 인덱스 맞춰주기 no 여기서 하면 음수일 때 값이 달라짐 그렇다고 abs 시키면 아래 if문 시행 못함
        if flow <= 0 :              # 음수 값이라면 나가는 것 #인덱스땜에 1빼줬으니 0포함시키기
            flow = abs(flow) -1
            idx = visit.index(abs(flow))        #해당 차가 주차되어있었던 공간 비어주기
            visit[idx] = 0       # 주차 공간 비워주기

        else :              # flow > 0 , 양수라면 들어오는 것
            flow -= flow    # 인덱스 맞춰주기 # 여기서 맞춰주면 1번 할때 0번이 되면서 에러 발생
            for i in range(N) :         # 앞부터 순서대로 확인
                if visit[i] == 0 :      # 비어있다면
                    visit[i] = flow      # 차 넣고 값 받기! 뺄 떄 필요하니 차번호로 저장,  역시 요금은 선불로
                    total += weight_fee[i] * car_weight[flow]
                    break               # 주차하면 중단해주기

    print(f'#{tc} {total}')

```

```python
# 2차시도 성공
# 주차 자리 공간이 없을 때 차가 온 것을 처리할 waiting을 만들었다



import sys

sys.stdin = open("test.txt")


'''
- 주차 가능한 공간 중 번호가 가장 작은 주차공간에 주차함
- 주차요금 : 차량의 무게 * 주차 공간마다 따로 책정된 단위 무게당 금액
- 진용이가 하루 벌어들일 총 수입은?
'''

'''
1. 주차장 공간 표시
2. 웨이팅 하는 차 표시
3. 차 빼고 작은 번호부터 다시 채우기
4. 인덱스 주의하기, 음수값 주의하기
'''


T = int(input())
for tc in range(1,T+1) :
    total = 0           #최종 답
    # 준비과정
    N,M = map(int,input().split())      # N : 1~N개의 주차 공간,주차 예정 차량 M대
    weight_fee = [0] * N                 # 인덱스 주의하기! 주차공간은 1번부터 시작, 인덱스는 0번부터
    for n in range(N) :
        weight_fee[n] = int(input())     # 단위 무게당 요금
    car_weight = [0] * M
    for m in range(M) :
        car_weight[m] = int(input())



    visit = [0] * N     # 차량의 주차 여부 기록
    waiting = []
    for j in range(2* M) :

        if waiting :
            for x in range(N) :
                if visit[x] == 0 :
                    total += weight_fee[x] * car_weight[waiting[0]-1]
                    visit[x] = waiting.pop(0)       #선입선출



        flow = int(input())      # 인덱스 맞춰주기 no 여기서 하면 음수일 때 값이 달라짐 그렇다고 abs 시키면 아래 if문 시행 못함
        if flow < 0 :              # 음수 값이라면 나가는 것
            idx = visit.index(abs(flow))        #해당 차가 주차되어있었던 공간 비어주기
            visit[idx] = 0       # 주차 공간 비워주기

        else :              # flow > 0 , 양수라면 들어오는 것
            for i in range(N) :         # 앞부터 순서대로 확인
                if visit[i] == 0 :      # 비어있다면
                    visit[i] = flow      # 차 넣고 값 받기! 뺄 떄 필요하니 차번호로 저장,  역시 요금은 선불로
                    total += weight_fee[i] * car_weight[flow-1] # 인덱스 맞춰주기
                    break               # 주차하면 중단해주기
            else :      # 주차 공간이 없을 때 대기 순번 등록
                waiting.append(flow)


    print(f'#{tc} {total}')

```
