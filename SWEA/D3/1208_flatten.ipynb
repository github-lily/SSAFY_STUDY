# Flatten 평탄화
- 높은 곳의 상자를 낮은 곳에 옮기는 방식으로 최고점과 최저점의 간격을 줄이는 작업
- 제한된 횟수만큼 옮기는 작업을 한 후 최고점과 최저점의 차이를 반환하라
- 테스트 케이스는 10개
- 가로 길이는 항상 100
- 상자의 높이는 항상 100 이하
- 덤프 횟수는 1이상 1000이하

```python
# 8.24 다시품

# Flatten

import sys

sys.stdin = open("test.txt", "r")

'''
- 높은 곳의 상자를 낮은 곳에 옮기는 방식으로 최고점과 최저점의 간격을 줄이는 작업을 수행
- 정해진 작업 횟수를 수행한 후 최고점과 최저점의 차이를 반환하는 프로그램
- 가로의 길이는 항상 100

'''

# 값 리셋 위치 항상 생각하기!!!!!!!!!!!!!!!!!!!!!!!!!
# 인덱스를 구해서 더하고 빼야 box 리스트에 반영이 됨

# max, min, index 안쓰고 구하기

for tc in range(1, 11):
    N = int(input())    #덤프 횟수
    box = list(map(int, input().split()))   #박스 높이


    # 덤프 횟수만큼 반복
    for _ in range(N) :
        # 최대, 최소 구하기
        # 값 리셋 위치 항상 생각하기!!!!!!!!!!!!!!!!!!!!!!!!!
        max_val = max_idx = min_idx = 0
        min_val = 9999999999999999
        for i in range(100) :

            # 최대
            if max_val < box[i] :
                max_val = box[i]
                max_idx = i


            # 최소
            if min_val > box[i] :
                min_val = box[i]
                min_idx = i


        # 덤프
        box[max_idx] -= 1
        box[min_idx] += 1
        

    # 덤프가 모두 끝난 후 최대 최소값 찾기
    # 값 리셋 위치 항상 생각하기!!!!!!!!!!!!!!!!!!!!!!!!!
    max_val = 0
    min_val = 999999999999
    for j in range(100) :

        # 최대
        if max_val < box[j] :
            max_val = box[j]


        # 최소
        if min_val > box[j] :
            min_val = box[j]


    ans = max_val - min_val
    print(f'#{tc} {ans}')
```

```python
# 간단하게 푸는 법

for tc in range(1, 11):
    N = int(input())    #덤프 횟수
    box = list(map(int, input().split()))   #박스 높이

    for _ in range(N) :
        max_idx = box.index(max(box))       #매번 바뀌니까 매번 새로 구하기
        min_idx = box.index(min(box))
        box[max_idx] -= 1
        box[min_idx] += 1

    ans = max(box) - min(box)
    print(f'#{tc} {ans}')
```

```python
'''
1. 최고점을 구해 -1, 최저점을 구해 +1을 한다
2. for 문을 이용해 이것을 range(정해진 덤프 횟수) 로 반복한다.
3. 반복한 뒤 변동된 최고점 최저점을 다시 구한다.
4. 차이를 구해 반환한다.
'''

for tc in range(1,11) :
    N = int(input())        #덤프 횟수
    arr = list(map(int,input().split()))

    #최고점과 최저점 구하기
    max_v = 0
    min_v = 101

    for i in range(100) :   #가로 길이 항상 100
        if arr[i] > max_v :
            max_v = arr[i]
        elif arr[i] < min_v :
            min_v = arr[i]
    
    # 덤프하기
    for x in range(N) :
        max_v -= 1
        min_v += 1

    # 변동 후 최고점 최저점 구하기
    c_max_v = 0
    c_min_v = 101

    for j in range(100) :
        if arr[j]>c_max_v :
            c_max_v = arr[j]
        elif arr[j] < c_min_v :
            c_min_v = arr[j]

    #차이 구하기
    diff = c_max_v - c_min_v

    print(f'#{tc} {diff}')

```

```python
#최고점 최저점 구하는 함수

def min_max(arr) :
    max_idx = min_idx = 0
    for i in range(100) :   #가로 길이 항상 100
        if arr[min_idx] > arr[i] :
            min_idx = i
        if arr[max_idx] < arr[i] :
            max_idx = i
    return min_idx, max_idx

#값 입력
for tc in range(1,11) :
    N = int(input())        #덤프 횟수
    arr = list(map(int,input().split()))

    #덤프하기   #덤프할 때마다 최대최소 구해야함!
    for x in range(N) :
        min_idx, max_idx  = min_max(arr)
        arr[min_idx] += 1
        arr[max_idx] -= 1

    #덤프 완료 후 최대 최소값 구하기
    min_idx, max_idx  = min_max(arr)
    print(f'#{tc} {arr[max_idx]- arr[min_idx]}')
    
```
