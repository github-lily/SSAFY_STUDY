- . : 비어있는 상태
- '#' : 막혀있는 상태
- 막혀있는 칸들이 **하나의**정사각형을 이루는지 판단(하나도 됨)

```python
# 1차 시도 : 오답
# 예외가 많음

T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for _ in range(N)]

    ans = "yes"

    for r in range(N) :
        cnt = 0
        for c in range(N) :
            # cnt 로 너비 구하기
            if arr[r][c] == '#' :
                cnt += 1
            if cnt >0 and arr[r][c] != '#' :
                break
            # #이 끝나면 행 확인
        if cnt>0 :
            for col in range(c, c-cnt, -1) :
                for row in range(r+1,r+cnt) :
                    if row >= N or col >= N :
                        ans = "no"
                        break
                    if arr[row][col] != '#' :
                        ans = "no"
                        break
            break




    print(f'#{tc} {ans}')
```

```python
# 2차 시도 : 정답
T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for _ in range(N)]

    ans = "yes"
    sr = sc =-1
    cnt = 0

    # '#' 찾고, 너비 구하기
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] == '#' :
                sr = r
                sc = c
                cnt +=1
                for width in range(sc+1,N) :
                    if arr[r][width] == '.' :
                        break
                    cnt += 1
                break
        if cnt > 0 :
            break


    # 영역 확인
    for row in range(sr+1,sr+cnt) :
        if ans == "no" : break  #답아니면 중단
        for col in range(sc, sc+cnt) :
            if row >= N or col >= N :
                ans = "no"
                break
            if arr[row][col] == '.' :
                ans = "no"
                break


    # 나머지 칸에 "#"이 있는지 확인
    for r in range(N) :
        for c in range(N) :
            if sr<= r <sr+cnt and sc <= c < sc+cnt :
                continue
            if arr[r][c] == '#' :
                ans = "no"
                break






    print(f'#{tc} {ans}')
```

```python
# 교수님 방법 복기!
# 정답! 실행시간과 메모리는 비슷함



def check(rs,cs,re,ce) :
    # 너비가 같은지 확인
    if (re-rs) != (ce-cs) :
        return False
    
    # 사각형 영역만큼 돌면서 모두 막혀있는지 확인
    for i in range(rs,re+1) :
        for j in range(cs,ce+1) :
            if arr[i][j] == '.' :
                return False
    return True


T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(input()) for _ in range(N)]

    # 정사각형의 좌상단, 우하단 좌표 찾기
    rs,cs = N,N     # 좌상단 : '#'의 최소 행, 열 값
    re,ce = 0,0     # 우하단 : '#'의 최대 행, 열 값
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] == '.' :
                continue
            # "#" 이라면 시작점과 끝점 찾기
            rs = min(r,rs)  # 좌상단
            cs = min(c,cs)  # 좌상단
            re = max(r,re)  # 우하단
            ce = max(c,ce)  # 우하단

    if check(rs,cs,re,ce) :
        print(f'#{tc} yes')
    else :
        print(f'#{tc} no')
```
