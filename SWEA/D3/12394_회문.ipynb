```python
# 1차 시도 : 실패

import sys

sys.stdin = open("test.txt", "r")


T = int(input())
for tc in range(1, T+1):
    N, M = map(int,input().split())  #배열의 크기      # 찾아야 하는 회문의 길이
    arr = [list(input()) for _ in range(N)]

    for r in range(N) :
        for c in range(N) :
            cnt = 0
            for i in range(M//2) :
                s = i
                e = M - s -1
                if c+e < N :
                    if arr[r][c+s] == arr[r][c+e] :
                        cnt += 1
                        if cnt == M//2 :
                            ans = arr[r][i:M-i]
                    else :
                        break


    print(f'#{tc} {ans}')
```

```python
# 교수님 코드

def find_palindrome(arr, N, M):
    cnt = 0
    # 모든 행에 대해
    for row in range(N):
        # arr[row] 행에 대해서 처리
        for s in range(0, N-M + 1):
            e = s + M - 1
            for i in range(M//2):
                # arr[s + i] <-> arr[e - i]
                if arr[row][s + i] != arr[row][e - i]:
                    break
            else:
                return arr[row][s: e + 1]
 
    # 모든 열에 대해
    for col in range(N):
        # arr[][col] 열에 대해서 처리
        for s in range(0, N-M + 1):
            e = s + M - 1
            for i in range(M//2):
                # arr[s + i] <-> arr[e - i]
                if arr[s + i][col] != arr[e - i][col]:
                    break
            else:
                ret = ''
                for i in range(s, e + 1):
                    ret += arr[i][col]
                return ret
 
    return None
 
T = int(input())
for tc in range(1, T + 1):
    N, M = map(int, input().split())
    arr = [input() for _ in range(N)]
 
    ans = find_palindrome(arr, N, M)
    print(f'#{tc} {ans}')
```

```python
# 복기

def find_s(arr, N, M) :
    # 행 먼저 탐색
    for row in range(N) :
        for s in range(N-M+1) :
            e = s+M-1
            for i in range(M//2) :
                if arr[row][s+i] != arr[row][e-i] :
                    break
            else :
                return arr[row][s:e+1]


    # 열 탐색
    for col in range(N) :
        for s in range(N-M+1) :
            e = s+M-1
            for i in range(M//2) :  #회문 길이의 반만큼 확인하면 됨
                if arr[s+i][col] != arr[e-i][col] :
                    break
            else :
                sen = ''
                for x in range(s,e+1) :
                    sen += arr[x][col]
                return sen

    return None



T = int(input())
for tc in range(1, T+1):
    N, M = map(int,input().split())  #배열의 크기      # 찾아야 하는 회문의 길이
    arr = [input() for _ in range(N)]   # 문자열도 인덱스 가능하니 굳이 리스트로 받을 필요 없음

    ans = find_s(arr,N,M)

    print(f'#{tc} {ans}')
```
