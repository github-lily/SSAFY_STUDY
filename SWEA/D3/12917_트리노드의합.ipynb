
# 트리_노드의합_확인용

- N개의 노드, 루트는 1번
- 자식노드의 합이 부모 노드
- 지정한 노드 번호에 저장된 값을 출력하라


```python
# 1차 시도 : 실패


def dfs(v) :
    if v > N :
        return

    dfs(v*2)
    if dfs(v//2) == 0 :
        tree[v//2] = dfs(v*2) + dfs(v*2+1)
    dfs(v*2+1)






T = int(input())
for tc in range(1,T+1) :
    N, M, L = map(int,input().split())

    tree = [0] * (N+1)
    # 리프노드의 값 저장
    for _ in range(M) :

        idx, num = map(int,input().split())

        tree[idx] = num

    dfs(1)
    print(tree[1:])
```

```python
# 9.1 2차 시도 : 성공

'''
[v*2] == 0 [v*2+1] == 0 : 단말노드라는 뜻
왼쪽과 오른쪽이 0이 아니고, v가 0일 때 
[v*2] + [v*2+1]
'''

T = int(input())
for tc in range(1,T+1) :
    # 노드의 개수, 리프 노드의 개수, 타겟 노드
    N, M, L = map(int,input().split())

    tree = [0] * (N+1)
    for _ in range(M) :
        # 리프 노드 번호와 값
        leaf_num, leaf_val = map(int,input().split())
        #단말노드에 저장
        tree[leaf_num] = leaf_val       


    # 아래부터 값 채우기(단말노드는 값을 아니까 패스)
    for v in range(N-M,0,-1) :
        if tree[v] == 0 :   #빈 값일 때
            if v*2 <= N and v*2+1 <= N :    #범위 이내일 때
                tree[v] = tree[v*2] + tree[v*2+1]
            # N번이라서 왼쪽 자식만 있을 경우
            elif v*2 == N :
                tree[v] = tree[v*2]

    print(f'#{tc} {tree[L]}')
```

```python
# 값이 덮어지면 오류가 날 수 있으므로 기존 값에 더하는 방식으로 진행
# 교수님 코드 1

T = int(input())
for tc in range(1, T + 1):
    N, M, L = map(int, input().split())
 
    tree = [0] * (N + 1)
    for _ in range(M):
        v, val = map(int, input().split())
        tree[v] = val
 
 
    for i in range(N-M, 0, -1):
        tree[i] += tree[i*2]
        if i*2 + 1 <= N:
            tree[i] += tree[i * 2 + 1]
 
    print(f'#{tc} {tree[L]}')
```

```python
# 교수님 코드 2

T = int(input())
for tc in range(1, T + 1):
    N, M, L = map(int, input().split())
 
    tree = [0] * (N + 1)
    for _ in range(M):
        v, val = map(int, input().split())
        tree[v] = val
 
    def dfs(v):
        if v > N:
            return 0
 
        l = dfs(v*2)
        r = dfs(v * 2 + 1)
 
        tree[v] += l + r
 
        return tree[v]
 
    dfs(1)
    print(f'#{tc} {tree[L]}')
```
