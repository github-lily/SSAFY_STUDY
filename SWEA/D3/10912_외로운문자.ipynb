```python
# 1차시도 (딕셔너리 활용)
# 1000개 중 164개 성공
# 값을 뒤에서부터 하나씩 적어야할 듯

'''
외로운 문자
소만자만으로 이루어진 문자열이 주어진다
이 문자열에서 같은 두 문자들을 짝짓고 남는 문자가 무엇인지 구하는 프로그램을 작성하라
같은 문자를 여러번 짝지어서는 안된다.
안남으면 Good을 출력

방법1
정렬, 앞에서부터 순회
i와 i+1이 같으면 삭제
남은거 출력

방법2
1. 딕셔너리로 풀기: key = 알파벳(str), value = 개수
1-1. aaaab -> { 'a':4, 'b':1 } -> 'a' 출력 x, 'b' 출력 O

방법3
스택

'''


T = int(input())
for tc in range(1,T+1) :
    text = input()

    N = len(text)


    cnt = {}
    ans = ''
    for char in text :
        cnt[char] = text.count(char)

        if cnt[char] % 2 == 1 :   # 짝이 없다면
            ans += char    # 리스트에 저장

    rest = sorted(ans)

    if ans == '' :  #문자열이 비어있다면
        print(f'#{tc} Good')

    else :
        answer = ''.join(rest)
        print(f'#{tc}', answer)
```

```python
# 4차시도 스택
import sys
sys.stdin = open("C:/Users/SSAFY/Desktop/0830/test.txt")

def s_push(item) :
    global top
    top += 1
    stack[top] = item

def s_pop() :
    global top
    top -= 1

T = int(input())
for tc in range(1,T+1) :
    text = input()
    N = len(text)

    sorted(text)
    ans = ''


    top = -1

    stack = [0] * (N+1)
    
    for i in text :
        s_push(i)
        if stack[top -1] == stack[top] :
            s_pop()
            s_pop()
        
    for x in stack :
        if x != 0 :
            ans += x
    
    if not stack :
        ans = "Good"

    print(ans)
```
