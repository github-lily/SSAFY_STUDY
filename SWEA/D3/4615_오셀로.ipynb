```python
# 1차 시도

'''

1 흑돌, 2 백돌
'''

dr = [0,1,1,1,0,-1,-1,-1]
dc = [1,1,0,-1,-1,-1,0,1]

T = int(input())
for tc in range(1,T+1) :
    N,M = map(int,input().split())      # N : 한 변의 길이 , M : 돌을 놓는 횟수
    arr = [[0] * N for _ in range(N)]

    # 초기 4개 돌 입력
    mid = N // 2
    arr[mid][mid] = arr[mid-1][mid-1] = 1
    arr[mid-1][mid] = arr[mid][mid-1] = 2


    # 돌 받기
    r = c = N // 2
    for _ in range(M) :
        pos = []
        r,c,dol = map(int,input().split())
        r = r - 1
        c = c - 1   #인덱스에 맞춰 값 변경
        for i in range(8) :
            for j in range(1,N) :
                nr = r + dr[i] * j
                nc = c + dc[i] * j
                if 0 <= nr < N and 0 <= nc < N :
                    if arr[nr][nc] == 0 :       # 돌이 아닐 때
                        break
                    if arr[nr][nc] == dol :     # 같은 색일 때
                        for a,b in pos :        # 다른 색일 때 저장해둔 값 빼오기
                            arr[a][b] = dol     # 돌 같은색으로 바꾸기
                        break    # 턴 끝남
                    else :                      # 다른 색일 때
                        pos.append([nr,nc])     # 나중에 돌 바꿀 수 있게 저장해두기

    B = W = 0
    for y in range(N) :
        for x in range(N) :
            if arr[y][x] == 1 :
                B += 1
            elif arr[y][x] == 2:
                W += 1

    print(f'#{tc} {B} {W}')

```

```python
# 교수님 코드 1

dr = [-1, -1, -1, 0, 1, 1, 1, 0]
dc = [-1, 0, 1, 1, 1, 0, -1, -1]
 
T = int(input())
for tc in range(1, T + 1):
    N, M = map(int, input().split())
    arr = [[0] * N for _ in range(N)]
 
    mid = N//2
    arr[mid-1][mid] = arr[mid][mid-1] = 1
    arr[mid][mid] = arr[mid-1][mid-1] = 2
 
    # 흑: 1, 백: 2
    for _ in range(M):
        r, c, dol = map(int, input().split())
        # 입력 좌표는 1부터 N사이 값 => 0에서 N-1사이의 값으로 변환
        r, c = r - 1, c - 1
 
        arr[r][c] = dol
 
        for i in range(8):
            # 그 방향으로 좌표값을 생성
            pos = []
            for j in range(1, N):
                nr = r + dr[i] * j
                nc = c + dc[i] * j
                if nr < 0 or nr >= N or nc < 0 or nc >= N:
                    break
                # arr[nr][nc] => 0, 1, 2 에 대해 처리
                if arr[nr][nc] == 0:
                    break
                if arr[nr][nc] == dol:
                    for a, b in pos:
                        arr[a][b] = dol
                    break
                else:
                    # 다른색의 돌의 위치를 저장
                    pos.append([nr, nc])
 
 
    B = W = 0
    for r in range(N):
        for c in range(N):
            if arr[r][c] == 1: B += 1
            elif arr[r][c] == 2: W += 1
 
    print(f'#{tc} {B} {W}')
    
```

```python
# 교수님 코드 2

dr = [-1, -1, -1, 0, 1, 1, 1, 0]
dc = [-1, 0, 1, 1, 1, 0, -1, -1]
 
T = int(input())
for tc in range(1, T + 1):
    N, M = map(int, input().split())
    arr = [[0] * N for _ in range(N)]
 
    mid = N//2
    arr[mid-1][mid] = arr[mid][mid-1] = 1
    arr[mid][mid] = arr[mid-1][mid-1] = 2
 
    # 흑: 1, 백: 2
    for _ in range(M):
        r, c, dol = map(int, input().split())
        # 입력 좌표는 1부터 N사이 값 => 0에서 N-1사이의 값으로 변환
        r, c = r - 1, c - 1
 
        arr[r][c] = dol
 
        for i in range(8):
            nr, nc = r + dr[i], c + dc[i]
            while 0 <= nr < N and 0 <= nc < N:
                if arr[nr][nc] == 0:
                    break
                if arr[nr][nc] == dol:
                    nr, nc = nr - dr[i], nc - dc[i]
                    while r != nr or c != nc:
                        arr[nr][nc] = dol
                        nr, nc = nr - dr[i], nc - dc[i]
                    break
                # =====================================
                nr, nc = nr + dr[i], nc + dc[i]
 
    B = W = 0
    for r in range(N):
        for c in range(N):
            if arr[r][c] == 1: B += 1
            elif arr[r][c] == 2: W += 1
 
    print(f'#{tc} {B} {W}')
```
