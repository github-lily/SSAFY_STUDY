# Magnetic

- 서로 다른 극이 붙은 곳(교착 상태)의 개수를 출력
- 100x100

```python
'''
0 은 빈 공간
1 은 N극
2 는 S극

위에 N 극(1)
아래에 S극 (2)

좌우 교착은 고려 안함

행렬을 탐색한다
    만약 arr[0][x]가 2라면 0으로 바꿔준다
    만약 arr[99][x]가 1이라면 0으로 바꿔준다
    
    값이 0이 아니라면
        아래쪽을 탐색한다
        아래 쪽에 0이 아닌 값이 있다면
            다른 값이면 ans +1을 한다
            i와 i+1 값을 모두 0으로 바꿔준다 (중복되지 않도록)
            같은 값이면 pass
            
    
방법2 : 교착이 2개일 때 해결이 안됨
위 아래를 탐색한다
    1이면 위를 탐색한다
        위쪽에 아무것도 없다면 값을 없앤다
        같은 값이 있다면 넘어간다
        다른 값이 있다면 ans + 1을 한다
            교착이 2개일 땐
            
    #위아래로 나누면 값이 중복될 수 있으므로 한 방향으로만 탐색한다        
    2이면 아래를 탐색한다
        아래쪽에 아무것도 없다면 값을 없앤다
        같은 값이 있다면 넘어간다
        다른 값이 있다면 ans +1을 한다
'''
```

```python
# 1차 시도
# xor 연산자 잘못 사용함
T = 10
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != 0 :     #값이 있을 때만 확인
                if arr[0][c] == 2 :
                    arr[0][c] = 0
                elif arr[99][c] == 1 :
                    arr[99][c] == 0
                else :
                    for x in range(1,N) :
                        nr = r + x
                        if r+x < N :    #범위 이내일 때 어차피 더하는 값이라 0 이상인지는 확인 안해도 됨
                            if arr[r][c]^arr[nr][c]  == 1 :
                            #같으면 0 출력 다르면 1 출력
                            #xor 연산자를 사용하면 값이 0 이어도 출력됨 다른걸로 인식됨 그래서 오류
                                ans += 1
                                arr[r][c] = arr[nr][c] = 0

    print(f'#{tc} {ans}')
```

```python
# 0이 아니고 값이 다를때로 조건을 바꾸고 
# 서로 다른 값을 찾으면  모두 0만 출력됨

T = 10
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != 0 :     #값이 있을 때만 확인
                if arr[0][c] == 2 :
                    arr[0][c] = 0
                elif arr[N-1][c] == 1 :
                    arr[N-1][c] = 0
                else :
                    for x in range(1,N) :
                        nr = r + x
                        if r + x < N :    #범위 이내일 때 어차피 더하는 값이라 0 이상인지는 확인 안해도 됨
                            if arr[nr][c] != 0 and arr[r][c]^arr[nr][c]  == 1 :
                            #같으면 0 출력 다르면 1 출력
                            #xor 연산자를 사용하면 값이 0 이어도 출력됨 다른걸로 인식됨 그래서 오류
                                ans += 1
                                arr[r][c] = arr[nr][c] = 0
                                break
                                # 1122 이 상황이어도 교착은 1이니까 더 안찾게 중간

    print(f'#{tc} {ans}')
```

```python
# 3차 시도
# xor 연산자 없애고 if 문 두개로 확인
# 값이 크게 나옴


T = 10
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != 0 :     #값이 있을 때만 확인
                if arr[0][c] == 2 : # s극이 맨위에 있다면 떨어져서 사라짐
                    arr[0][c] = 0
                elif arr[N-1][c] == 1 :  #n극이 맨 아래에 있다면 떨어져서 사라짐
                    arr[N-1][c] = 0
                else :
                    for x in range(1,N) :
                        nr = r + x
                        if r + x < N :    #범위 이내일 때 어차피 더하는 값이라 0 이상인지는 확인 안해도 됨
                            if arr[r][c] == 1 :
                                if arr[nr][c] == 2 :
                                    ans +=1
                                    arr[r][c] = arr[nr][c] = 0  #1212 이 상황일 떄 교착이 3개가 되지 않도록 초기화
                                    break    # 1122 이 상황이어도 교착은 1개니까 더 안찾게 중간
                            if arr[r][c] == 2:
                                if arr[nr][c] == 1:
                                    ans += 1
                                    arr[r][c] = arr[nr][c] = 0
                                    break


    print(f'#{tc} {ans}')
```

```python
# 교수님 코드

import sys; sys.stdin = open('1220.txt')

for test_case in range(1, 11):
    SIZE = int(input())
    arr = [list(input().split()) for _ in range(100)]

    ans = 0
    for c in range(SIZE):
        state = 0
        for r in range(SIZE):
            if state == 0 and arr[r][c] == '1':
                state = 1
            elif state == 1 and arr[r][c] == '2':
                ans += 1
                state = 0

    print(f'#{test_case} {ans}')

```

```python
# gpt 코드

T = 10
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(map(int, input().split())) for _ in range(N)]

    ans = 0
    for c in range(N):
        stack = []
        for r in range(N):
            if arr[r][c] == 1:  # N극이면 stack에 추가
                stack.append(arr[r][c])
            elif arr[r][c] == 2:  # S극을 만났을 때
                if stack:  # stack에 N극이 있을 경우 교착 발생
                    ans += 1
                    stack = []  # stack 초기화 (교착 상태이므로)

    print(f'#{tc} {ans}')
```

```python
# gpt 코드 복기

for tc in range(1, 11):
    N = int(input())
    arr = [list(map(int,input().split())) for _ in range(N)]

    ans = 0
    for c in range(N) :
        stack = []      #한 열을 모두 순회하면 리셋
        for r in range(N) :
            if arr[r][c] == 1 : #N극일 떄 stack에 저장
                stack.append(1)
            elif arr[r][c] == 2 : #S 극이 있을 때
                if stack :      #stack이 빈리스트가 아니라면
                    ans += 1
                    stack = []      #리스트 초기화
                    
    print(f'#{tc} {ans}')
```
