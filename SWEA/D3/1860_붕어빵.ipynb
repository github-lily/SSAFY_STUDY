# 1860 진기의 최고급 붕어빵
- N : 예약 가능한 사람의 수
- 0초 부터 붕어빵 만들기 시작
- M초 : K개의 붕어빵을 만드는데 걸리는 시간
- 기다리는 시간 없이 붕어빵을 줄 수 있으면 possible, 안되면 impossible
- people : 사람들이 도착하는 시간 초


```python
# 1차 시도
# 누적값이 반영이 안됨
# ex) 2초에 2개를 굽고 3초에 손님이 와도 rest[3] 은 0 이기 떄문에 impossible이 되버림



T = int(input())
for tc in range(1,T+1):
    N,M,K = map(int,input().split())    #N : 사람수 / M : K개 붕어빵 만드는데 걸리는 시간
    time = sorted(list(map(int,input().split())))

    rest = [0] * (time[-1] + M)       # 마지막 손님이 오는 시간까지 잔여 개수 카운트 # 넉넉하게 M 더해주기

    ans = "possible"        #중간에 안되는 상황이 안걸리면 그대로 출력
    x_range = round(time[-1]/M)       #나눈 값 반올림

    for x in range(1,x_range+1) :      #마지막 손님이 오는 시간을 M초로 나눈 것만큼 반복
        rest[M*x] += K
    for t in time :
        if rest[t] <= 0 :
            ans = "impossible"      #잔여가 없으면 impossible로 바뀌고 멈춤
            break
        rest[t] -= 1


    print(f'#{tc} {ans}')

```

```python
# 2차 시도
# for문이 불필요하게 다 돌아감,,
# 값이 대부분 possible이 나와버림

T = int(input())
for tc in range(1,T+1):
    N,M,K = map(int,input().split())    
    # N : 사람수 / M : K개 붕어빵 만드는데 걸리는 시간
    time = sorted(list(map(int,input().split())))

    ans = "possible"  # 중간에 안되는 상황이 안걸리면 그대로 출력
    rest = 0

    for k in range(1, 11112) :
        if k % M == 0 :     # M초가 지났을 때
            rest += K
        for t in time :
            if k == t :     #손님이 온 시간인데
                if rest <= 0 :  #잔여가 없다면
                    ans = "impossible"      #불가능 반환
                    break       #반복 멈춤
                else :
                    rest -= 1
            if ans == "impossible" :
                break
            break   #for문이 모두 돌면 멈춤


    print(f'#{tc} {ans}')

```

```python
# 승윤이 코드

T = int(input())
for tc in range(1,T+1):
    N,M,K = map(int,input().split())
    # N : 사람수 / M : K개 붕어빵 만드는데 걸리는 시간
    time = sorted(list(map(int,input().split())))

    ans = "possible"  # 중간에 안되는 상황이 안걸리면 그대로 출력


    for i in range(N):
        fish = time[i] // M * K
        # print(fish, time[i] )
        fish -= (i + 1)

        if fish < 0:
            ans = 'Impossible'


    print(f'#{tc} {ans}')

```

```python
# 영광이 코드 + 주석

# 모든 손님들에게 기다리는 시간없이 붕어빵을 제공할 수 있는지 여부를 판단하는 함수
def boongeobbang(N, M, K, arrive_time):
    arrive_time.sort()      # 손님 오는 시각 오름차순 정렬
    cur_sec = M             # 현재 시각 (초기값은 한번 구운 상태)
    cur_num = K             # 현재 남아 있는 붕어빵 개수 (초기값은 한번 구운 상태)
    for guest_sec in arrive_time:
        if cur_sec < guest_sec:                     # 아직 손님 오기 전이면
            while cur_sec + M <= guest_sec:         # 손님 올 때까지 계속 구워  #현재시간+붕어빵굽는 시간 <= 손님오는 시간
                cur_sec += M                        # 현재시간 늘려주기
                cur_num += K                        # 현재 붕어빵 수 늘려주기
        if guest_sec - M < 0 or cur_num == 0:       # 너무 일찍 오거나 왔는데 붕어빵이 없으면
            return 'Impossible'                     # 못 줘
        cur_num -= 1                                # 손님 한명 왔으니 붕어빵 하나 줘
    return 'Possible'                               # 다 줬으면 성공
 
for tc in range(1, 1+int(input())):
    # N: 손님 수, M: 븡어빵 K개 만드는데 걸리는 시간, K: M초만에 만들 수 있는 붕어빵 수
    N, M, K = map(int, input().split())
    arrive_time = list(map(int, input().split()))       # 손님이 도착하는 시각
    answer = boongeobbang(N, M, K, arrive_time)
    print(f'#{tc}', answer)
```
