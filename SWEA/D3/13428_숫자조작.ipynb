- 9자리 이하의 정수 N
- 위치를 한 번만 바꿔 새로운 수 M을 만들 수 있다(한번도 안 바꿀수도 있음)
- M의 최솟값과 최댓값을 구하는 프로그램
- M의 맨 앞에 '0'이 나타나면 안됨

```python
import sys
sys.stdin = open("test.txt")

'''
- 완전탐색 no (N 이 999,999,999이하)
- 그리디 (제일 큰 값을 앞으로 옮기면 최댓값, 제일 작은 값을 앞으로 옮기면 최솟값)
- 한번만 바꾸는거라 가능!
- 맨 앞 값과 뒤부터 탐색해서 제일 큰 값이랑 교환

'''
T = int(input())
for tc in range(1,T+1) :
    num = input()
    N = len(num)


    a = max(num)
    b = min(num)


    # 역순으로 돌면서 최댓값 최솟값의 인덱스 구하기
    # 같이 돌리면 브레이크가 애매해짐
    for max_idx in range(N-1,-1,-1) :
        if num[max_idx] == a :
            break

    for min_idx in range(N-1,-1,-1) :
        if b == '0':
            b = 1
        if num[min_idx] == b :
            if min_idx == 0 :
                # 맨 앞 숫자가 가장 작을 땐..? 2를 어떻게찾지? 또 반복?
                break


    max_val = list(num)
    max_val[0], max_val[max_idx] = max_val[max_idx],max_val[0]

    min_val = list(num)
    min_val[0], min_val[min_idx] = min_val[min_idx], min_val[0]


    min_val = ''.join(min_val)
    max_val = ''.join(max_val)

    print(f'#{tc}', min_val, max_val)




```

```python
# 계속 수정해도 문제생김...!
import sys
sys.stdin = open("test2.txt")



'''
- 완전탐색 no (N 이 999,999,999이하)
- 그리디 (제일 큰 값을 앞으로 옮기면 최댓값, 제일 작은 값을 앞으로 옮기면 최솟값)
- 한번만 바꾸는거라 가능!
- 맨 앞 값과 뒤부터 탐색해서 제일 큰 값이랑 교환

'''
T = int(input())
for tc in range(1,T+1) :
    num = input()
    N = len(num)

    front = 0       # 나중에 자리 바꿀 때 사용
    a = max(num)
    b = min(num)


    # 역순으로 돌면서 최댓값 최솟값의 인덱스 구하기
    # 같이 돌리면 브레이크가 애매해짐
    for max_idx in range(N-1,-1,-1) :
        if num[max_idx] == a :
            break

    for min_idx in range(N-1,-1,-1) :
        if b == '0':
            b = 1
        if num[min_idx] == b :
            # 만약 맨 앞이 최솟값이면 다음 최소값 찾기
            if min_idx == 0 :
                b = int(b)+1
                front = 1
                    # 맨 앞이 최소인 경우 두번째 값과 바꿈
                    # 그럼 두번째 값이 또 최소이면..?
            else :
                break



    max_val = list(num)
    max_val[front], max_val[max_idx] = max_val[max_idx],max_val[front]

    min_val = list(num)
    min_val[front], min_val[min_idx] = min_val[min_idx], min_val[front]


    min_val = ''.join(min_val)
    max_val = ''.join(max_val)


    print(f'#{tc}', min_val, max_val)




```

```python
# 3차시도 : 성공

import sys
sys.stdin = open("test2.txt")



T = int(input())
for tc in range(1,T+1) :
    num = list(input())
    N = len(num)

    # 값 저장해두기
    min_val = max_val = value = int(''.join(num))


    for i in range(N-1) :
        for j in range(i+1,N) :
            # 만약 맨 앞과 바꿀 자리가 0이라면 넘어감
            if i == 0 and num[j] == '0'  :
                continue
            num[i],num[j] = num[j],num[i]
            value = int(''.join(num))
            # 다음 비교를 위해 원상복구
            num[i], num[j] = num[j], num[i]

            min_val = min(min_val,value)
            max_val = max(max_val,value)


    print(f'#{tc}', min_val, max_val)




```
