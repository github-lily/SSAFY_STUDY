1번부터 N번까지의 번호가 부여된 N개의 물건
최대 부피 K만큼 넣을 수 있는 가방

1번 물건부터 N번 물건 각각은 부피 V와 가치 C를 가지고 있다.

물건들 중 몇개를 선택하여 가방에 넣어서 얻을 수 있는 최대 가치는?

```python
# 1차시도 (백트레킹, 시간초과)



def backtracking(s, total, val) :       # 부피 total, 가치 val
    global max_val

    if total > K :
        return

    if s == N :
        if max_val < val :
            max_val = val
        return

    for i in range(N) :
        # 포함 안시킨거
        backtracking(s+1,total,val)

        # 포함 시킨거
        if visit[i] == 0 :
            visit[i] = 1
            backtracking(s+1,total+gold[i][0],val+gold[i][1])
            visit[i] = 0





T = int(input())
for tc in range(1, T+1):
    N, K = map(int,input().split())
    gold = []
    max_val = 0
    visit = [0] * N

    for _ in range(N) :
        V, C = map(int,input().split())
        gold.append([V,C])

    backtracking(0,0,0)

    print(f'#{tc} {max_val}')
```

```python
# 2차 시도
# 가지치기 추가(확인했던건 확인 안하기)


def backtracking(s, total, val, start) :       # 부피 total, 가치 val
    global max_val



    if total > K :
        return

    if s == N :
        if max_val < val :
            max_val = val
        return


    for i in range(N) :

        # 포함 안시킨거
        backtracking(s+1,total,val,i+1)

        # 포함 시킨거
        if visit[i] == 0 :
            visit[i] = 1
            backtracking(s+1,total+gold[i][0],val+gold[i][1], i+1)
            visit[i] = 0




T = int(input())
for tc in range(1, T+1):
    N, K = map(int,input().split())
    gold = []
    max_val = 0
    visit = [0] * N


    for _ in range(N) :
        V, C = map(int,input().split())
        gold.append([V,C])

    backtracking(0,0,0,0)

    print(f'#{tc} {max_val}')
```
