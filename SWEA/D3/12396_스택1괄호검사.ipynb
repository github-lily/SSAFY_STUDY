# 12396 스택1_괄호검사

- 괄호가 제대로 들어갔는지 확인하는 프로그램
- 정상적인 짝인 경우 1, 아닌 경우 0

- 여는 괄호가 나오면 스택에 넣고
- 닫는 괄호가 나오면 하나씩 스택에서 빼면서 짝이 맞는지 검사
- 괄호 오류
  - 여는 괄호가 없는 경우 (스택이 빈 경우)
  - 닫는 괄호가 없는 경우 (스택이 남은 경우)
  - 괄호의 종류가 다른 경우 (비교했을 때 다른 경우)

```python
T = int(input())
for tc in range(1, T + 1):
    arr = list(input())

    N = len(arr)
    s = [0] * N
    top = -1


    for i in range(len(arr)):

        if arr[i] == '(' or arr[i] == '{':
            top += 1
            s[top] = arr[i]
        elif arr[i] == ')' or arr[i] == '}':
            if s[top] != arr[i]:  # 괄호의 종류가 다른 경우
                result = 0
            elif top == -1:  # 스택이 빈 경우
                result = 0
            elif s[top] == arr[i]:  # 정상인 경우 top pop
                top -= 1
    if top == -1:       #반복을 모두 돈 후 리스트가 비었다면
        result = 1

    print(f'#{tc} {result}')


    # 이렇게하면 괄호의 짝을 확인할 수 없음!!! (와 )는 다르고, (와 }는 다름!! 각각 조건!
```
