# 5215. 햄버거 다이어트


정해진 칼로리 이하의 조합 중 민기가 가장 선호하는 햄버거를 조합해주는 프로그램 만들기
같은 재료는 재사용 불가
주어진 칼로리 이하의 조합 중 점수가 가장 높은 햄버거의 점수
 
완전탐색
모든 조합을 구하고
각각의 점수와 칼로리를 계산한다
만약 제한 칼로리 이상 음식이있으면 제거한다(가지치기)
 
백트레킹
칼로리의 합이 L를 넘지 않는 모든 조합을 구한 뒤
만족도를 계산한다
 


```python
DP, DFS로도 풀어보기
```

```python
# 1차 시도
# 런타임 에러

def backtrack(k,n,case) :
    #현재 상태를 저장(공집합도)
    idx_lst.append(case[:])
 
    for j in range(k,n) :
        case.append(idx[j])
        backtrack(j+1,n, case)
        case.pop()
 
 
 
 
T = int(input())
for tc in range(1,T+1) :
    N, L = map(int, input().split())
    # 재료의 수, 제한 칼로리
    best = 0            # 1등
 
    idx = [x for x in range(N)]
    idx_lst = []
 
 
    score = [0] * N     # 재료 점수
    kcal = [0] * N      # 재료 칼로리
 
    for i in range(N) :
        S, K = map(int, input().split())
        # 재료에 대한 점수와 칼로리
 
        score[i] = S
        kcal[i] = K
 
    backtrack(0, N, [])
 
 
    for lst in idx_lst :
        kcal_sum = 0
        score_sum = 0
        for a in lst :
            kcal_sum += kcal[a]
            if kcal_sum <= L :
                score_sum += score[a]
                if best < score_sum :
                    best = score_sum
 
    print(f'#{tc} {best}')
```

```python
# 2차시도
# 런타임 에러

def backtrack(k,n,case) :
    #현재 상태를 저장(공집합도)
    idx_lst.append(case[:])
 
    for j in range(k,n) :
        case.append(idx[j])
        backtrack(j+1,n, case)
        case.pop()
 
 
 
 
T = int(input())
for tc in range(1,T+1) :
    N, L = map(int, input().split())
    # 재료의 수, 제한 칼로리
    best = 0            # 1등
 
    idx = [x for x in range(N)]
    idx_lst = []
 
 
    score = [0] * N     # 재료 점수
    kcal = [0] * N      # 재료 칼로리
 
    for i in range(N) :
        S, K = map(int, input().split())
        # 재료에 대한 점수와 칼로리
 
        score[i] = S
        kcal[i] = K
 
    backtrack(0, N, [])
 
 
    for lst in idx_lst :
        # 칼로리가 제한보다 낮을 떄
        if sum(kcal[k] for k in lst) <= L :
            # 점수를 구하기
            sum_val = sum(score[s] for s in lst)
            # 최고점수 갱신하기
            if best < sum_val :
                best = sum_val
 
    print(f'#{tc} {best}')
```

```python
# 정답

def backtrack(k,n,case) :
    global best
 
 
    # 칼로리가 제한보다 낮을 떄
    if sum(kcal[k] for k in case) <= L :
        # 점수를 구하기
        sum_val = sum(score[s] for s in case)
 
        # 최고점수 갱신하기
        if best < sum_val :
            best = sum_val
 
    for j in range(k,n) :
        case.append(j)
        backtrack(j+1,n, case)
        case.pop()
 
 
 
 
T = int(input())
for tc in range(1,T+1) :
    N, L = map(int, input().split())
    # 재료의 수, 제한 칼로리
    best = 0            # 1등
 
 
 
    score = [0] * N     # 재료 점수
    kcal = [0] * N      # 재료 칼로리
 
    for i in range(N) :
        S, K = map(int, input().split())
        # 재료에 대한 점수와 칼로리
 
        score[i] = S
        kcal[i] = K
 
    backtrack(0, N, [])
 
 
 
 
    print(f'#{tc} {best}')
```
