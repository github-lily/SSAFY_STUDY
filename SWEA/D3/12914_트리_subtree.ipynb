```python
import sys
sys.stdin = open("test.txt")


from collections import deque


#너비우선탐색 사용해보기

def bfs(start) :
    cnt = 1     #자기 자신 포함
    q = deque()
    q.append(start)



    while q :
        node = q.popleft()

        # 왼쪽 확인
        if left[node] != 0 :
            q.append(left[node])
            cnt+=1

        # 오른쪽 확인
        if right[node] != 0 :
            q.append(right[node])
            cnt+=1

    return cnt



T = int(input())
for tc in range(1,T+1) :
    E, R = map(int,input().split())
    # 노드 번호는 1번부터 E+1번까지 주어짐
    # R 루트 N 정점 E 간선

    N = E+1

    arr = list(map(int,input().split()))

    left = [0] * (N + 1)
    right = [0] * (N +  1)
    par = [0] * (N + 1)
    # 1번부터 있으니 +1을 해야함

    for i in range(E) :
        p,c = arr[i*2], arr[i*2+1]


        if left[p] == 0 :
            left[p] = c

        else :
            right[p] = c
        par[c] = p

    ans = bfs(R)
    print(f'#{tc} {ans}')
```
