```python
import sys
sys.stdin = open("test.txt")


from collections import deque


#너비우선탐색 사용해보기

def bfs(start) :
    cnt = 1     #자기 자신 포함
    q = deque()
    q.append(start)



    while q :
        node = q.popleft()

        # 왼쪽 확인
        if left[node] != 0 :
            q.append(left[node])
            cnt+=1

        # 오른쪽 확인
        if right[node] != 0 :
            q.append(right[node])
            cnt+=1

    return cnt



T = int(input())
for tc in range(1,T+1) :
    E, R = map(int,input().split())
    # 노드 번호는 1번부터 E+1번까지 주어짐
    # R 루트 N 정점 E 간선

    N = E+1

    arr = list(map(int,input().split()))

    left = [0] * (N + 1)
    right = [0] * (N +  1)
    par = [0] * (N + 1)
    # 1번부터 있으니 +1을 해야함

    for i in range(E) :
        p,c = arr[i*2], arr[i*2+1]


        if left[p] == 0 :
            left[p] = c

        else :
            right[p] = c
        par[c] = p

    ans = bfs(R)
    print(f'#{tc} {ans}')
```

```python
# 9.1 복기

import sys
sys.stdin = open("test.txt")

'''
노드 N을 루트로 하는 서브 트리에 속한 노드의 개수 구하기
'''

'''
1. 부모 자식의 노드 쌍을 리스트에 저장한다
2. N의 자식 노드를 탐색하며 cnt+= 1
3. cnt 를 출력한다
'''

def find_root(r) :
    global cnt

    if r == 0 :
        return

    cnt += 1
    find_root(left[r])
    find_root(right[r])

T = int(input())
for tc in range(1,T+1) :
    # 간선 개수와 루트 노드
    E,R = map(int,input().split())
    arr = list(map(int,input().split()))
    N = E+1
    cnt = 0

    # 계층구조를 나타낼 리스트
    left = [0] * (N+1)
    right = [0] * (N+1)
    par = [0] * (N+1)

    # 계층구조에 담기
    for i in range(E) :
        p,c = arr[i*2], arr[i*2+1]

        # 왼쪽 자식이 없다면
        if left[p] == 0 :
            left[p] = c
        # 왼쪽 자식이 있다면 (이진트리라 왼쪽부터 넣기 떄문에 왼쪽은 없고 오른쪽은 있는 경우는 없음)
        else :
            right[p] = c
        par[c] = p      # c의 부모는 p이다

    find_root(R)

    print(f'#{tc} {cnt}')
```
