```python
# 1차 시도
# 풀다가 for 문이 너무 많아져서 포기함

T = int(input())
for tc in range(1,T+1) :
    N = int(input)
    arr = [list(input().split()) for i in range(N)]

    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != '.' :   #돌일 때 확인
                # 오른쪽 확인
                cnt = 1     #본인 돌 포함
                for k in range(1, 4) :
                    if arr[r][c+k] == '.' or c+k < 0 or c+k >= N :    #돌이 아니거나 범위 밖이라면
                        break
                    else :
                        cnt += 1
                        if cnt == 5 :
                            ans = 'YES'

                # 아래 확인
                for k in range(1,4) :
                    if arr[r+k][c] == '.' or r+k < 0 or r+k >= N :
                        break
                    else :
                        cnt +=1
                        if cnt == 5 :
                            ans = 'YES'
                            




    print(f'#{tc} {ans}')
```

```python
# 2차 시도
# for 문 최대한 줄여서 코드 간결화
# 문제에 빈칸으로 구분 없으므로 split 제거함
# 다 NO 나옴


'''
오른쪽과 아래, 우측아래, 좌측 아래 대각선만 확인하기
돌이 5개가 되기 전 .을 만나면 break 하기


'''


T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for i in range(N)]

    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != '.' :   #돌일 때 확인
                cnt = 1     #본인 돌 포함
                for k in range(1, 4) :
                    # 오른쪽
                    # 범위 조건을 한번에 확인하면 아래는 되고 우측은 안되는 상황에서 아래도 못보고 넘어갈 수 있음
                    if 0 <= c+k < N :     # 범위일 때
                        if arr[r][c+k] == '.' : #돌이 아니라면
                            continue    #실행안하고 건너뜀
                    # 아래
                    if 0<= r+k <N :
                        if arr[r+k][c] == '.' :
                            continue
                    # 우측대각선
                    if 0<= r+k < N and 0<= c+k <N :
                        if arr[r+k][c+k]  == '.' :
                            continue
                    # 좌측 대각선
                    if 0<= r+k < N and 0<= c-k <N :
                        if arr[r+k][c-k] == '.' :
                            continue
                    else :
                        cnt += 1
                        if cnt == 5 :
                            ans = 'YES'
    else : ans = 'NO'


    print(f'#{tc} {ans}')
```

```python
# 3차 시도
# continue, pass, if, elif 모두 써봐도 if문에서 걸리면 k가 2로 넘어가버림
# 한 for 문 안에 있어 그렇다는걸 깨달음
# 포기하고 델타 시도(4차 시도)

import sys

sys.stdin = open("test2.txt", "r")


'''

오른쪽과 아래, 우측아래, 좌측 아래 대각선만 확인하기
돌이 5개가 되기 전 .을 만나면 break 하기


'''


T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for i in range(N)]

    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != '.' :   #돌일 때 확인
                cnt = 1     #본인 돌 포함
                for k in range(1, 5) :
                    # 오른쪽
                    # 범위 조건을 한번에 확인하면 아래는 되고 우측은 안되는 상황에서 아래도 못보고 넘어갈 수 있음
                    if 0 <= c+k < N :     # 범위일 때
                        if arr[r][c+k] == '.' : #돌이 아니라면
                            cnt = 0
                    # 아래
                    elif 0<= r+k <N :
                        if arr[r+k][c] == '.' :
                            continue
                    # 우측대각선
                    if 0<= r+k < N and 0<= c+k <N :
                        if arr[r+k][c+k]  == '.' :
                            continue
                    # 좌측 대각선
                    if 0<= r+k < N and 0<= c-k <N :
                        if arr[r+k][c-k] == '.' :
                            continue
                    else :
                        cnt += 1
                        if cnt == 5 :
                            ans = 'YES'
    else : ans = 'NO'


    print(f'#{tc} {ans}')
```

```python
# 4차 시도
# 정답
# 가지치기 해야함
# 돌이 아니면 for 문 중단하는거
# 오목을 찾으면 아예 나머지를 확인 안하는 것
# 남은 칸이 4칸이 안되면 아예 확인을 안하는 것


T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for i in range(N)]
    ans = 'NO'      #찾으면 YES로 바뀜 못찾으면 NO 그대로 출력
 
    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != '.' :   #돌일 때 확인
                dr = [0,1,1,1]      #우, 우대각, 하, 좌대각
                dc = [1,1,0,-1]
                for i in range(4) :
                    cnt = 1     #한 방향에서 5개가 쭉 있어야하므로 여기서 리셋 #내 돌까지 포함이니까 1부터 시작
                    for j in range(1,5) :
                        nr = r + dr[i] * j
                        nc = c + dc[i] * j
                        if 0<= nr < N and 0 <= nc < N :     #범위 이내일 때
                            if arr[nr][nc] == '.' :     #돌이 아니라면
                                cnt = 1     # 누적값 초기화
                                break       #탐색 중단
                            else :  #돌이라면
                                cnt += 1
                                if cnt == 5 :
                                    ans = 'YES'     #찾았을 때 YES로 바꿈
                                    break
                        else : #범위가 아니어도
                            cnt = 1 #누적값 초기화
                            break #탐색 중단
 
 
 
 
    print(f'#{tc} {ans}')
```

```python
# 5차 시도
# 남은 칸이 4칸이 안되면 아예 확인을 안하는 것 시도했는데 실패함
# r의 범위를 0부터 N-4 ,  N-3 으로 변경해봤지만 100개 중 2개의 테스트케이스에서 오답이 나옴
# (이 경우 실행시간은 160, 180으로 빨라지긴 함)


import sys

sys.stdin = open("test2.txt", "r")


'''
오른쪽과 아래, 우측아래, 좌측 아래 대각선만 확인하기
돌이 5개가 되기 전 .을 만나면 break 하기


'''


T = int(input())
for tc in range(1,T+1) :
    N = int(input())
    arr = [list(input()) for i in range(N)]
    ans = 'NO'      #찾으면 YES로 바뀜 못찾으면 NO 그대로 출력

    for r in range(N) :
        for c in range(N) :
            if arr[r][c] != '.' :   #돌일 때 확인
                dr = [0,1,1,1]      #우, 우대각, 하, 좌대각
                dc = [1,1,0,-1]
                for i in range(4) :
                    # 가지치기
                    if i == 0 : #오른쪽 확인할 때
                        if c >= N-4 : #오른쪽에 남은 칸이 4개가 안될 때
                            break
                    elif i == 2 :   #아래 확인할 때
                        if r >= N-4 :    #아래 남은 칸이 4개가 안될 때
                            break
                    else :
                        cnt = 1     #한 방향에서 5개가 쭉 있어야하므로 여기서 리셋 #내 돌까지 포함이니까 1부터 시작
                        for j in range(1,5) :
                            nr = r + dr[i] * j
                            nc = c + dc[i] * j
                            if 0<= nr < N and 0 <= nc < N :     #범위 이내일 때
                                if arr[nr][nc] == '.' :     #돌이 아니라면
                                    cnt = 1     # 누적값 초기화
                                    break       #탐색 중단
                                else :  #돌이라면
                                    cnt += 1
                                    if cnt == 5 :
                                        ans = 'YES'     #찾았을 때 YES로 바꿈
                                        break
                            else : #범위가 아니어도
                                cnt = 1 #누적값 초기화
                                break #탐색 중단




    print(f'#{tc} {ans}')
```

```python
# 6차 시도
# if문을 이용해 답을 찾으면 for문을 중단하도록 함
# 실행시간 191ms 에서 변동 없음

import sys
sys.stdin = open("test.txt","r")


T = int(input())
for tc in range(1, T + 1):
    N = int(input())
    arr = [list(input()) for i in range(N)]
    ans = 'NO'  # 찾으면 YES로 바뀜 못찾으면 NO 그대로 출력

    for r in range(N):
        for c in range(N):
            if arr[r][c] != '.':  # 돌일 때 확인
                dr = [0, 1, 1, 1]  # 우, 우대각, 하, 좌대각
                dc = [1, 1, 0, -1]
                for i in range(4):
                    cnt = 1  # 한 방향에서 5개가 쭉 있어야하므로 여기서 리셋 #내 돌까지 포함이니까 1부터 시작
                    for j in range(1, 5):
                        nr = r + dr[i] * j
                        nc = c + dc[i] * j
                        if 0 <= nr < N and 0 <= nc < N:  # 범위 이내일 때
                            if arr[nr][nc] == '.':  # 돌이 아니라면
                                cnt = 1  # 누적값 초기화
                                break  # 탐색 중단
                            else:  # 돌이라면
                                cnt += 1
                                if cnt == 5:
                                    ans = 'YES'  # 찾았을 때 YES로 바꿈
                                    break

                        else:  # 범위가 아니어도
                            cnt = 1  # 누적값 초기화
                            break  # 탐색 중단
                    if ans == 'YES' :   # 이미 찾았으면 탐색 중단
                        break
                if ans == 'YES':  # 이미 찾았으면 탐색 중단
                    break
        if ans == 'YES':  # 이미 찾았으면 탐색 중단
            break


    print(f'#{tc} {ans}')
```
