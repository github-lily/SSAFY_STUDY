### 문제 설명
- 이진 트리에서 임의의 두 정점의 가장 가까운 공통 조상을 찾고, 그 정점을 루트로 하는 서브 트리의 크기를 알아내는 프로그램
- 서브트리의 크기엔 공통 조상도 포함됨
- 정점 번호는 1부터 V까지의 정수, 루트 정점은 항상 1번
- 시간제한 : 파이썬 30초, 메모리 256MB

### 입력값
1. 테스트케이스 수
2. 정점의 개수 V, 간선의 개수 E, 공통 조상을 찾는 두개의 정점 번호
3. E개의 간선 나열 (부모, 자식 순서)

### 출력값
#t 가장가까운조상의번호 서브트리의크기(정점개수)

```python
# 메모리 49,236kb, 시간 204ms
# 한번에 성공!!
# 부모자식관계를 같이 저장하면 좀 더 빠를듯!
# => 리스트 안에 리스트..!?

from collections import deque

# N1의 부모를 리스트에 저장하는 함수
def find_N1_parents(start) :
    global N1_parents

    q = deque()
    v = [0] * (V+1)

    q.append(start)
    v[start] = 1
    N1_parents.add(start)

    while q :
        c = q.popleft()
        for w in parents[c] :
            if v[w] == 0 :
                q.append(w)
                v[w] = 1
                N1_parents.add(w)


# N1 부모 set과 N2의 부모를 비교하는 함수
def find_parents(start) :
    global ancestor
    
    q = deque()
    v = [0] * (V+1)

    q.append(start)
    v[start] = 1

    while q :
        c = q.popleft()

        # N1의 조상에 있는 값이라면 종료
        if c in N1_parents :
            ancestor = c
            return
        
        for w in parents[c] :
            if v[w] == 0 :
                q.append(w)
                v[w] = 1


# 자식 트리 개수 세는 함수
def count_tree(start) :
    global cnt

    q = deque()
    v = [0] * (V+1)

    q.append(start)
    v[start] = 1

    while q :
        c = q.popleft()
        
        for w in children[c] :
            if v[w] == 0 :
                q.append(w)
                v[w] = 1
                cnt += 1


T = int(input())
for tc in range(1,T+1) :
    V, E, N1, N2= map(int,input().split())
    temp = list(map(int,input().split()))
    ancestor = 0
    cnt = 1

    # 부모관계만 저장하는 리스트(공통조상찾기용)
    parents = [[] for _ in range(V+1)]

    # 자식관계만 저장하는 리스트(트리크기측정용)
    children = [[] for _ in range(V+1)]

    for p in range(0,E*2,2) :
        parents[temp[p+1]].append(temp[p])
        children[temp[p]].append(temp[p+1])
    

    # N1의 부모를 리스트에 저장하는 함수
    N1_parents = set()
    find_N1_parents(N1)

    
    # N1 부모 set과 N2의 부모를 비교하는 함수
    find_parents(N2)


    # 자식 트리 개수 세는 함수
    count_tree(ancestor)


    # 답 출력
    print(f'#{tc} {ancestor} {cnt}')


```
