```python
# 2차 배열 4분할
N = 5
arr = [[0] * N for _ in range(N)]

# 배열을 출력하는 함수
def print_arr():
    for line in arr:
        print(*line)
    print('---------------')


# 배열을 리셋하는 함수
def clear_arr():
    arr = [[0] * N for _ in range(N)]


# 영역을 구하는 함수
# 시작 인덱스 기준
# 시작행, 시작열, 끝행, 끝열, 입력값
def set_area(sr, sc, er, ec, val):
    for r in range(sr, er):
        for c in range(sc, ec):
            arr[r][c] = val


for i in range(1, N):
    for j in range(1, N):
        # 1. 좌상단:(0, 0) -> 우하단:(i-1, j-1)
        # r -> range(0, i), c -> range(0, j)
        set_area(0, 0, i, j, '-')   # 1구간(좌상단)
        # 2. (0, j) -> (i-1, N-1)   # 0,0~i-1,j-1
        set_area(0, j, i, N, '+')   # 2구간(우상단)
        # 3. (i, 0) -> (N-1, j-1)   # 0,j-1 ~ i,N-1
        set_area(i, 0, N, j, '*')   # 3구간(좌하단)
        # 4. (i, j) -> (N-1, N-1)   # i,0 ~ N-1,j-1
        set_area(i, j, N, N, '#')   # 4구간(우하단)

        print_arr()
        clear_arr()
```
