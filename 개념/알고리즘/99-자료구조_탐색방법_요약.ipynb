자료구조와 탐색 방법에 대한 특징 요약 정리

# 자료구조

### List
**1. 배열**
- 일정한 자료형의 변수들을 하나의 이름으로 열거하여 사용하는 자료구조
- 하나의 선언을 통해 둘 이상의 변수를 선언할 수 있다는 장점

**2. 정렬**  
1) 버블 정렬
- 인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식
- 시간복잡도 : O(n²)
- 코드가 단순하지만 연산량이 많아 시간이 오래걸린다

2) 카운팅 정렬
- 항목들의 순서를 결정하기 위해 집합에 각 항목이 몇 개씩 있는지 세는 작업을 하여 선형 시간에 정렬하는 효율적인 알고리즘
- 숫자 카운트, 카운트 각각의 합을 이용해 리스트를 정렬
- 정수 또는 정수로 표현할 수 있는 자료형에서만 적용 가능
- 시간복잡도 : O(n+k)
- n이 비교적 작을 때만 가능하다. 연산량이 적지만 코드가 더 복잡하다

3) 선택 정렬
- 주어진 자료들 중 가장 작은 값의 원소부터 차례대로 선택하여 위치를 교환하는 방식
- 교환의 횟수가 버블, 삽입정렬보다 작다

- 셀렉션 알고리즘
    - 저장되어 있는 자료로부터 k번째로 크거나 작은 원소를 찾는 방법
    - 최소값, 최대값 혹은 중간값을 찾는 알고리즘을 의미하기도 함
    - 자료 정렬하기 -> 원하는 순서에 있는 원소 가져오기

4) 퀵
- 주어진 배열을 두 개로 분할하고 각각을 정렬한다

5) 삽입

6) 병합


**3. 순열**
- 서로 다른 것들 중 몇 개를 뽑아서 한 줄로 나열하는 것
- 서로 다른 n개 중 r개를 택하는 순열은 `nPr` 같이 표현한다. 
- nPr = n * (n-1) * (n-2)*...*(n-r+1) = 곱하는 수의 개수가 r개가 될 떄까지 곱한다.
- nPn = n! (Factorial)

### String

#### 매칭에 사용되는 검색 알고리즘

1. 고지식한 패턴검색 알고리즘
- 본문 문자열을 처음부터 끝까지 차례대로 순회하면서 패턴 내의 문자들을 일일이 비교하는 방식으로 동작

2. 카프-라빈 알고리즘

3. KMP 알고리즘(B형 대비)
- 불일치가 발생한 텍스트 스트링의 앞 부분에 어떤 문자가 있는지를 미리 알고 있으므로, 불일치가 발생한 앞 부분에 대하여 다시 비교하지 않고 매칭을 수행
- 즉, 매칭이 실패했을 때 돌아갈 곳을 e미리 계산하여 그곳부터 다시 시작함

4. 보이어-무어 알고리즘(B형 대비)
- 패턴의 오른쪽 끝 글자와 비교하여 일치하지 않는다면 패턴만큼 건너뛰는 것


### STACK
- 선형구조(자료 간의 관계가 1대 1의 관계를 갖는 것)
- 후입선출(LIFO) : 마지막에 삽입한 자료가 가장 먼저 나오는 것
- pop/push


### Queue
- 뒤에서는 삽입만, 앞에서는 삭제만 이루어지는 삽입과 삭제의 위치가 제한적인 자료구조
- 선입선출구조(FIFO) : 가장 먼저 삽입된 원소는 가장 먼저 삭제된다.



# 탐색 방법

**1. 완전 탐색**
- 문제의 해결법으로 생각할 수 있는 모든 경우의 수를 구해서 확인하는 방법
- 일반적으로 경우의 수가 상대적으로 작을 때 유용하다
- 정확성은 높지만 효율성은 낮다(속도가 느림)


**2. 탐욕(Greedy) 기법**
- 여러 경우 중 하나를 결정해야 할 때마다 그 순간에 최적이라고 생각되는 것을 선택해 나가는 방식
- 각 선택은 근시안적으로는 최적이지만, 최종 해답이 최적이라는 보장은 없음

**3. 순차 검색**
- 가장 간단하고 직관적인 방법
- 순차 구조로 구현된 자료구조에서 원하는 항목을 찾을 때 유용하다
- 검색 대상의 수가 많은 경우 수행 시간이 급격히 증가하여 비효율적이다
- 정렬되지 않은 경우 / 정렬된 경우(찾는 값보다 더 큰 값이 나오면 멈추기 때문에 검색 실패를 반환하는 경우 평균 비교 회수가 반으로 줄어든다)

**4. 이진 검색**
- 자료의 가운데에 있는 항목의 키 값과 비교하여 다음 검색의 위치를 결정하고 검색을 계속 진행하는 방법
- 자료가 정렬된 상태에서 적용할 수 있음
- 자료에 삽입이나 삭제가 발생했을 때 배열의 상태를 새로 정렬해야 하는 추가 작업이 필요


**5. 해쉬**

**6. DP(dynamic Programming)**
- 입력 크기가 작은 부분 문제들을 모두 해결한 후 그 해들을 이용하여 보다 큰 크기의 부분 문제들을 해결하여, 최종적으로 원래 주어진 입력의 문제를 해결하는 알고리즘
- 즉, 문제를 부분 문제로 나누고 가장 작은 부분 문제부터 해를 구한 뒤 그 결과를 테이블에 저장하고, 저장된 부분 문제의 해를 이용하여 상위 문제의 해를 구한다
- Memoization : 이전에 계산한 값을 메모리에 저장해서 매번 다시 계산하지 않도록 하는 방법. 전체적인 실행 속도를 빠르게 하는 기술이다.


**7. DFS(깊이우선탐색)**
- 스택이나 재귀 호출을 이용하여 구현할 수 있다
- 시작 노드에서부터 한 경로를 따라 최대한 깊게 탐색한 후 더이상 갈 곳이 없으면 가장 마지막에 있었던 갈림길의 정점으로 되돌아와서 다른 경로를 탐색하는 방법
- 탐색을 계속 반복하여 결국 모든 정점을 방문하는 순회방법
- 스택 사용 : 방문한 노드를 스택에 저장하고, 더 이상 탐색할 수 없을 때 스택에서 노드를 꺼내어 역순으로 출력한다
- 그래프 구조 파악 및 미로 찾기 문제에 유용하다

- 백트래킹
    - 해를 찾는 도중 막히면 되돌아가서 다시 해를 찾아 가는 기법
    - 가지치기 : 유망하지 않은 노드가 포함되는 경로는 더 이상 고려하지 않는 것
    - 최적화문제, 결정 문제(미로찾기 등 조건을 만족하는 해가 존재하는지 여부를 확인하는 문제)에 적용 가능


**8. BFS(너비우선탐색)**
- 큐를 이용하여 구현한다
- 시작 노드에서부터 인접한 정점들을 모두 탐색한 후 다음 정점으로 이동한다.
- 탐색을 진행하면서 방문한 노드를 큐에 저장하고, 먼저 저장된 노드부터 출력한다. 
- 최단 경로 또는 최단 거리 등을 구하는데 활용된다.
