# 알고리즘


#### 알고리즘 표현
- 컴퓨터 분야에서 알고리즘을 표현하는 방법은 크게 2가지
- 의사코드(슈도코드, Pseudocode)와 순서도
</br>



#### 무엇이 좋은 알고리즘인가?
- 이건 기본 ㅎㅎ 시험에 안나옴
1. 정확성 : 얼마나 정확하게 동작하는가
2. 작업량 : 얼마나 적은 연산으로 원하는 결과를 얻어내는가
3. 메모리 사용량 : 얼마나 적은 메모리를 사용하는가
4. 단순성 : 얼마나 단순한가
5. 최적성 : 더 이상 개선할 여지 없이 최적화 되었는가


#### 알고리즘의 성능
- 많은 문제에서 성능 분석의 기준으로 알고리즘의 작업량을 비교한다



- 1 과 100, 2와 99 ... 를 더하면 모두 101이 됨 그러므로 101 x 100, 2개씩 나오니까 나누기 2
</br>
- 알고리즘 작업량은 시간복잡도로 표현한다
- 시간 복잡도(Time Complexity)
    - 실제 걸리는 시간을 측정
    - 실행되는 명령문의 개수를 계산
-  시간복잡도 : 빅-오(O) 표기법
    - 빅-오 표기법(Big-O Notation)
    - 시간 복잡도 함수 중에서 가장 큰 영향력을 주는 n에 대한 항만을 표시(가장 높은 차수를 나타냄)
    - 계수(Coefficient)는 생략하여 표시
    - 예 : n 개의 데이터를 입력 받아 저장한 후 각 데이터에 1씩 증가시킨 후 각 데이터를 화면에 출력하는 알고리즘의 시간 복잡도는? 
        - O(n)

- 배열의 필요성
    - 프로그램 내에서 여러 개의 변수가 필요할 때, 일일이 다른 변수명을 이용하여 자료에 접근하는 것은 매우 비효율적일 수 있다.
    - 배열을 사용하면 하나의 선언을 통해서 둘 이상의 변수를 선언할 수 있다.
    - 단순히 다수의 변수 선언을 의미하는 것이 아니라, 다수의 변수로는 하기 힘든 작업을 배열을 활용해 쉽게 할 수 있다.
